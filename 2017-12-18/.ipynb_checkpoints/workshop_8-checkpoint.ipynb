```python
from larlib import *
 
#definisco qualche colore
cLegno = [0.3,0.2,0.1]
cTegola = [0.6,0.05,0]
cMattone = [1,0.95,0.8]
cErba = [0.2,0.4,0.05]
cPietra = [0.7,0.7,0.72]
cLegnoChiaro=[0.627451,0.321569,0.176471]


#metodo che mi serve per creare un vetro trasparente
def hex_material(color, light, trasparence):

    def hex_to_rgb(value):
        value = value.lstrip('#')
        lv = len(value)
        return map(lambda x: x/255., list(int(value[i:i + lv // 3], 16) for i in range(0, lv, lv // 3)))

    params = hex_to_rgb(color) + [.1] + \
             hex_to_rgb(light) + [trasparence] +\
             hex_to_rgb(light) + [.1] +\
             hex_to_rgb("#000000") + [.1] +\
             [100]

    return MATERIAL(params)

window = CUBOID([1,0.1,1])
window = hex_material("#ffffff", "#ffffff", .4)(window)
#VIEW(window)
```

```python
#in questo metodo genero una soporgenza sopra alla finestra
def pilastro_finestra_sopra(x,y,z):
    
    x_pilastro=x
    y_pilastro=y
    z_pilastro=z

    # definisco i vertici
    V=[[0,0,0],#1 quadrato dx
    [0,y_pilastro,0],#2 quadrato dx
    [0,y_pilastro,z_pilastro],#3 quadrato dx
    [0,0,z_pilastro],#4 quadrato dx
    [0,y_pilastro*1.5,z_pilastro],#5 triangolo dx
    [x_pilastro,0,0],#6 quadrato sx
    [x_pilastro,y_pilastro,0],#7 quadrato sx
    [x_pilastro,y_pilastro,z_pilastro],#8 quadrato sx
    [x_pilastro,0,z_pilastro],#9 quadrato sx
    [x_pilastro,y_pilastro*1.5,z_pilastro],#10 triangolo sx
    [0,0,-z_pilastro],#11 sotto dx
    [0,y_pilastro,-z_pilastro],#12 sotto dx
    [x_pilastro,0,-z_pilastro],#13 sotto sx
    [x_pilastro,y_pilastro,-z_pilastro]#14 sotto sx
    ]
    
    #collego i vari vertici
    pilastro_rettangolo=MKPOL([V,[[1,4,3,5,2],[6,9,8,10,7]],[]])
    pilastro_rettangolo_cubo=MKPOL([V,[[1,2,11,12],[6,7,13,14]],[]])
    pilastro_rettangolo=JOIN(pilastro_rettangolo)
    pilastro_rettangolo_cubo=JOIN(pilastro_rettangolo_cubo)
    pilastri=STRUCT([pilastro_rettangolo,pilastro_rettangolo_cubo])
    return pilastri

#VIEW(pilastro_finestra_sopra(1,0.5,0.1))

#Metodo che crea una finestra. Composto da: due ante, 4 pezzi di vetro, due pali(uno verticale e uno orizzontale), una sporgenza e due pali sopra e sotto
#tip=0 sporgenza lunga 
#tip=1 sporgenza piu corta a destra
#tip=2 sporgenza piu corta a sinistra
#tip=3 sporgenza piccola
#tetto=0 senza tetto
#tetto=1 con tetto
def finestra(x,y,z,tip,tetto):
    #creo le varie componenti per la finestra
    z_pali=z*0.05
    palo_sopra=COLOR(cLegno)(CUBOID([x,y*0.5,z_pali]))
    
    y=y*0.3
    palo_sotto=COLOR(cLegno)(CUBOID([x,y,z_pali]))
    
    x_palo_orizzontale=x
    z_palo_orizzontale=z*0.05
    palo_orizzontale=COLOR(cLegno)(CUBOID([x_palo_orizzontale,y,z_palo_orizzontale]))
    
    x_palo_verticale=x*0.05
    z_palo_verticale=z*2.95
    palo_verticale=COLOR(cLegno)(CUBOID([x_palo_verticale,y,z_palo_verticale]))
    
    x_vetro_sopra=x*0.45
    z_vetro_sopra=z*1.15
    vetro_sopra=hex_material("#ffffff", "#ffffff", .4)(CUBOID([x_vetro_sopra,y,z_vetro_sopra]))
    
    x_vetro_sotto=x*0.45
    z_vetro_sotto=z*1.7
    vetro_sotto=hex_material("#ffffff", "#ffffff", .4)(CUBOID([x_vetro_sotto,y,z_vetro_sotto]))
    
    x_anta=x*0.05
    y_anta=y*2.3
    z_anta=z*3
    ante=COLOR(cLegno)(CUBOID([x_anta,y_anta,z_anta]))
    #unisco il tutto
    window=STRUCT([ante,T(1)(x_anta)(palo_sotto),
                   T([1,3])([x_anta,z_pali])(vetro_sotto),
                   T([1,3])([x_anta+x_vetro_sotto,z_pali])(palo_verticale),
                   T([1,3])([x_anta+x_vetro_sotto+x_palo_verticale,z_pali])(vetro_sotto),
                   T(1)(x_anta+x_vetro_sotto+x_palo_verticale+x_vetro_sotto)(ante),
                   T([1,3])([x_anta,z_pali+z_vetro_sotto])(palo_orizzontale),
                   T([1,3])([x_anta,z_pali+z_vetro_sotto+z_palo_orizzontale])(vetro_sopra),
                   T([1,3])([x_anta+x_vetro_sopra+x_palo_verticale,z_pali+z_vetro_sotto+z_palo_orizzontale])(vetro_sopra),
                   T([1,3])([x_anta,z_pali+z_vetro_sotto+z_palo_orizzontale+z_vetro_sopra])(palo_sopra)])
    
    #creo il contorno per la finestra
    y_contorno=y_anta*0.9
    contorno=CUBOID([x*1.3,y_contorno,z_anta+(x*1.3-SIZE(1)(window))/2])
    contorno=DIFFERENCE([contorno,T(1)((SIZE(1)(contorno)-SIZE(1)(window))/2)(S(2)(y*10)(window))])
    contorno=STRUCT([window,T(1)(-(SIZE(1)(contorno)-SIZE(1)(window))/2)(contorno)])
    contorno=COLOR(cPietra)(contorno)
    window=STRUCT([window,contorno])
    x_window=SIZE(1)(window)
    
    #aggiungo la sporgenza sotto in base a quella che mi serve
    z_sporgenza=z*0.05
    if tip==0:   
        sporgenza=COLOR(GRAY)(CUBOID([x*4.1,y*1.6,z_sporgenza]))
        window=BOTTOM([window,sporgenza])
        
    if tip==1:
        sporgenza=COLOR(GRAY)(CUBOID([x*2.9,y*1.6,z_sporgenza]))
        window=STRUCT([window,T([1,2,3])([-x*0.3,(SIZE(2)(window)-y*1.6)/2,-z_sporgenza])(sporgenza)])
    
    if tip==2:
        sporgenza=COLOR(GRAY)(CUBOID([x*2.9,y*1.6,z_sporgenza]))
        window=STRUCT([window,T([1,2,3])([-x*1.5,(SIZE(2)(window)-y*1.6)/2,-z_sporgenza])(sporgenza)])
        
    if tip==3:
        sporgenza=COLOR(GRAY)(CUBOID([x*1.2,y*1.6,z_sporgenza]))
        window=BOTTOM([window,sporgenza])
    
    #aggingo il tetto piccolo sopra alla finestra utilizzando il metodo creato in precedenza
    if tetto==1:
        z_sopra=z*0.15
        sopra=COLOR(cPietra)(CUBOID([x*1.4,y_contorno*1.1,z_sopra]))
        window=STRUCT([window,T([1,3])([-(x*1.4-x_window)/2-x*0.25/2,SIZE(3)(window)-z_sporgenza])(sopra)])
        pilastro=pilastro_finestra_sopra(x*1.8,y_contorno*0.5,z_sopra/2)
        pilastro=COLOR(cPietra)(T([1,2,3])([-z_sopra-x*0.25,y_contorno*0.7,SIZE(3)(window)-z_sporgenza+z_sopra/2])(pilastro))
        window=STRUCT([window,pilastro])
    
    return window
#VIEW(finestra(1.3,1,1,3,1))
```

```python
#metodo che crea un altra tipologia di finestra fatta da due pezzi di vetro un anta verticale e un contorno di legno e uno in pietra
def finestrella(x,y,z):
    
    y=y*0.3
    x_contorno,z_contorno=x,z*1.5
    
    #creo il contorno in pietra
    contorno=CUBOID([x_contorno,y*1.95,z_contorno])
    contorno_pietra=CUBOID([x_contorno*1.3,y*1.8,z_contorno*1.3])
    contorno_pietra=DIFFERENCE([contorno_pietra,T([1,3])([(x_contorno*1.3-x_contorno)/2,(z_contorno*1.3-z_contorno)/2])(contorno)])
    contorno_pietra=COLOR(cPietra)(contorno_pietra)
    x_dif_contorno,z_dif_contorno=x_contorno-0.1,z_contorno-0.1
    dif_contorno=CUBOID([x_dif_contorno,y*1.95,z_dif_contorno])
    
    #creo la cornice in legno
    cornice=DIFFERENCE([contorno,T([1,3])([(x_contorno-x_dif_contorno)/2,(z_contorno-z_dif_contorno)/2])(dif_contorno)])
    cornice=COLOR(cLegno)(cornice)
    
    #creo le varie componenti della finestra
    x_palo=x*0.05
    palo=CUBOID([x_palo,y,z_dif_contorno])
    x_vetro,z_vetro=(x_dif_contorno-x_palo)/2,z_dif_contorno
    vetro=CUBOID([x_vetro,y,z_vetro])
    palo=COLOR(cLegno)(palo)
    vetro=hex_material("#ffffff", "#ffffff", .4)(vetro)
    vetro=STRUCT([vetro,T(1)(x_vetro)(palo),T(1)(x_vetro+x_palo)(vetro)])
    
    #assemblo il tutto
    finestra=STRUCT([cornice,T([1,3])([(x_contorno-x_dif_contorno)/2,(z_contorno-z_dif_contorno)/2])(vetro),
                   T([1,2,3])([-(x_contorno*1.3-x_contorno)/2,y*0.2,-(z_contorno*1.3-z_contorno)/2])(contorno_pietra)])
    return finestra

#VIEW(finestrella(1.3,1,1))
```

```python
#tip==0 mattoncino con bordo
#tip==1 mattoncino senza bordo laterale
#tip==2 mattoncino senza bordo superiore ed inferiore
def mattoncino(arg,tip):
    x,y,z=arg
    if z<x:
        spessore_contorno_x=z*0.2
        spessore_contorno_z=z*0.2
    else:
        spessore_contorno_x=x*0.2
        spessore_contorno_z=x*0.2
    if tip==1:
        spessore_contorno_x=0
    if tip==2:
        spessore_contorno_z=0
    
    contorno=CUBOID([x,y*0.8,z])
    centro=CUBOID([x-spessore_contorno_x,y,z-spessore_contorno_z])
    contorno=DIFFERENCE([T(2)((y-y*0.9)/2)(contorno),T([1,3])([(spessore_contorno_x)/2,(spessore_contorno_z)/2])(centro)])
    cubo=STRUCT([T(2)((y-y*0.9)/2)(contorno),T([1,3])([(spessore_contorno_x)/2,(spessore_contorno_z)/2])(centro)])
    
    return cubo
#VIEW(COLOR(cMattone)(mattoncino([1.3*0.8,0.5,0.3625],0)))
```

```python
#arg i valori x,y,z
#numero_mattoni_larghezza numero di mattoni per orizzontale
#numero_mattoni_altezza numero di mattoni per verticale
def mattonata(arg,numero_mattoni_larghezza,numero_mattoni_altezza):
    x,y,z=arg
    x=x/numero_mattoni_larghezza
    z=z/numero_mattoni_altezza
    mattone=COLOR(cMattone)(mattoncino([x,y,z],0))
    muro=STRUCT(NN(numero_mattoni_larghezza)([mattone,T(1)(x)]))
    muro=STRUCT(NN(numero_mattoni_altezza)([muro,T(3)(z)]))
    return muro
#VIEW(mattonata([10,0.5,1.8],10,4))
```

```python
# crea una maniglia per le porte
# utilizzo 4 curve di bezier e poi le assemblo tra loro
def maniglia(x,y,z):
    x=x/1.3
    curva_sx=BEZIER(S1)([[x*0,y*0,z*0],[x*0,y*2.4,z*1],[x*0,y*2.4,z*2],[x*0,y*2.4,z*3],[x*0,y*0,z*4.4]])
    curva_up=BEZIER(S1)([[x*0.2,y*0,z*0.2],[x*0.2,y*2.2,z*1],[x*0.2,y*2.2,z*2],[x*0.2,y*2.2,z*3],[x*0.2,y*0,z*4.2]])
    curva_2D=BEZIER(S2)([curva_sx,curva_up]) 
  
    curva_dx=BEZIER(S1)([[-x*0.2,y*0,z*0.2],[-x*0.2,y*2.2,z*1],[-x*0.2,y*2.2,z*2],[-x*0.2,y*2,z*3],[-x*0.2,y*0,z*4.2]])
    curva_down=BEZIER(S1)([[x*0,y*0,z*0.4],[x*0,y*2,z*1],[x*0,y*2,z*2],[x*0,y*2,z*3],[x*0,y*0,z*4]])
    curva_2D1=BEZIER(S2)([curva_dx,curva_down]) 
    
    dominio1=PROD([INTERVALS(1)(32),INTERVALS(1)(1)])

    curva_3D=BEZIER(S3)([curva_2D,curva_2D1]) 
    dominio3d=PROD([dominio1,QUOTE([1])])
    curva_3D=MAP(curva_3D)(dominio3d)
    return COLOR(cLegnoChiaro)(curva_3D)

#VIEW(maniglia(1.3,1,1))
```

```python
# un altra tipologia di finestra fatta da un vetro un contorno ed è aperta sotto
#tip==0 è una finestra con la parte sotto chiusa da un contorno in pietra
#tip==1 è una finestra con la parte sotto aperta
#con_maniglia==0 senza maniglia
#con_maniglia==1 con maniglia
def finestrella_sotto(x,y,z,tip,con_maniglia):
    x_contorno,z_contorno=x,z*1.2
    
    # creo il contorno della finestra
    contorno=CUBOID([x_contorno,y*0.6,z_contorno]) 
    x_dif_contorno,z_dif_contorno=x_contorno-(x*0.1),z_contorno-(z*0.1)
    z_p=z_contorno/3.0
    y_p=y*0.4
    
    #creo il contorno in pietra con piastrelle di varie dimensioni
    p_medio=COLOR(cMattone)(mattoncino([x*0.4,y_p,z_p],0))
    p_piccolo=COLOR(cMattone)(mattoncino([x*0.2,y_p,z_p],0))
    p_grande=mattonata([x*1.8,y_p,z*0.3],5,1)
    y_spostamento=y*0.2
    piastrella1=T([1,2])([x_contorno,y_spostamento])(p_piccolo)
    piastrella2=T([1,2,3])([x_contorno,y_spostamento,z_p])(p_medio)
    piastrella3=T([1,2,3])([x_contorno,y_spostamento,z_p*2])(p_piccolo)
    piastrella4=T([1,2,3])([-x*0.4,y_spostamento,z_p*3])(p_grande)
    piastrella5=T([1,2])([-x*0.2,y_spostamento])(p_piccolo)
    piastrella6=T([1,2,3])([-x*0.4,y_spostamento,z_p])(p_medio)
    piastrella7=T([1,2,3])([-x*0.2,y_spostamento,z_p*2])(p_piccolo)
    piastrelle=STRUCT([piastrella1,piastrella2,piastrella3,piastrella4,piastrella5,piastrella6,piastrella7])
    
    #il tip==0 è una finestra con la parte sotto chiusa da una pietra
    if tip==0:
        dif_contorno=CUBOID([x_dif_contorno,y*0.6,z_dif_contorno])
        
        #creo le varie componenti
        cornice=DIFFERENCE([contorno,T([1,3])([(x_contorno-x_dif_contorno)/2,(z_contorno-z_dif_contorno)/2])(dif_contorno)])
        cornice=COLOR(cLegno)(cornice)
        vetro=hex_material("#ffffff", "#ffffff", .4)(CUBOID([x_dif_contorno,y*0.3,z_dif_contorno]))
        piastrella=T([1,2,3])([-x*0.4,y_spostamento,-z*0.3])(p_grande)
        piastrelle=STRUCT([piastrella,piastrelle])
        
        finestra=STRUCT([cornice,piastrelle,
                         T([1,3])([(x_contorno-x_dif_contorno)/2,(z_contorno-z_dif_contorno)/2])(vetro)])
    
    #il tip==1 è una finestra con la parte sotto aperta
    else:
        y=y*0.5

        #creo componenti per la porta
        palo_sopra=COLOR(cLegno)(CUBOID([x,y,z*0.05]))
        z_palo_sotto=z*0.15
        palo_sotto=COLOR(cLegno)(CUBOID([x,y,z_palo_sotto]))

        x_palo_verticale=x*0.05
        z_palo_verticale=z*1.3
        palo_verticale=COLOR(cLegno)(mattoncino([x_palo_verticale,y*1.01,z_palo_verticale],0))
        x_palo_verticale=x*0.1

        if x<z:
            x_anta=x*0.1
            x_anta_alt=z*0.1
        else:
            x_anta=z*0.1
            x_anta_alt=x*0.1

        x_vetro=x*0.35+x_anta_alt-x_anta
        z_vetro=z_palo_verticale
        vetro=hex_material("#ffffff", "#ffffff", .4)(CUBOID([x_vetro,y,z_vetro]))


        z_anta=z_vetro
        ante=COLOR(cLegno)(CUBOID([x_anta,y,z_anta]))

        #assemblo la porta
        finestra=STRUCT([palo_sotto,T(3)(z_palo_sotto)(ante),
                       T([1,3])([x_anta,z_palo_sotto])(vetro),
                       T([1,3])([x_anta+x_vetro,z_palo_sotto])(palo_verticale),
                       T([1,3])([x_anta+x_vetro+x*0.05,z_palo_sotto])(palo_verticale),
                       T([1,3])([x_anta+x_vetro+x_palo_verticale,z_palo_sotto])(vetro),
                       T([1,3])([x_anta+x_vetro+x_palo_verticale+x_vetro,z_palo_sotto])(ante),
                       T(3)([z_palo_sotto+z_anta])(palo_sopra)])
        p_medio=COLOR(cMattone)(mattoncino([x*0.4,y_p,z*0.3],0))
        piastrelle=T(3)(z*0.3)(piastrelle)
        piastrella=STRUCT([T([1,2,3])([x_contorno,y_spostamento])(p_medio),T([1,2,3])([-x*0.4,y_spostamento])(p_medio)])
        piastrelle=STRUCT([piastrella,piastrelle])
        
        #assemblo la finestra con le piastrelle e con la maniglia
        finestra=STRUCT([piastrelle,finestra])
        if con_maniglia==1:
            maniglie=maniglia(x/14.0,y/8,z/14.0)
            maniglie_dietro=R([1,2])(PI)(maniglie)
            finestra=STRUCT([finestra,
                       T([1,2,3])([x_vetro+x_anta+x*0.02,y,z*0.6])(maniglie),
                       T([1,2,3])([x_vetro+x*0.08+x_anta,y,z*0.6])(maniglie),
                       T([1,3])([x_vetro+x_anta+x*0.02,z*0.6])(maniglie_dietro),
                       T([1,3])([x_vetro+x*0.08+x_anta,z*0.6])(maniglie_dietro)
                      ])
    return finestra

#VIEW(finestrella_sotto(1.3,1,1,1,1))
#VIEW(finestrella_sotto(1.3,1,1,0,0))
```

```python
# metodo che crea una porta con un anta verticale due pezzi di vetro e un contorno di legno
# tip==1 con contorno
# tip==0 senza contorno
# con_maniglia==1 con maniglia
# con_maniglia==0 senza maniglia
def porta(x,y,z,tip,con_maniglia):
    y=y*0.5
    
    #creo componenti per la porta
    palo_sopra=COLOR(cLegno)(CUBOID([x,y,z*0.05]))
    z_palo_sotto=z*0.15
    palo_sotto=COLOR(cLegno)(CUBOID([x,y,z_palo_sotto]))
      
    x_palo_verticale=x*0.05
    z_palo_verticale=z*2.8
    palo_verticale=COLOR(cLegno)(mattoncino([x_palo_verticale,y*1.01,z_palo_verticale],0))
    x_palo_verticale=x*0.1
    
    if x<z:
        x_anta=x*0.1
        x_anta_alt=z*0.1
    else:
        x_anta=z*0.1
        x_anta_alt=x*0.1
    
    x_vetro=x*0.35+x_anta_alt-x_anta
    z_vetro=z_palo_verticale
    vetro=hex_material("#ffffff", "#ffffff", .4)(CUBOID([x_vetro,y,z_vetro]))
    
    
    z_anta=z_vetro
    ante=COLOR(cLegno)(CUBOID([x_anta,y,z_anta]))
    
    #assemblo la porta
    door=STRUCT([palo_sotto,T(3)(z_palo_sotto)(ante),
                   T([1,3])([x_anta,z_palo_sotto])(vetro),
                   T([1,3])([x_anta+x_vetro,z_palo_sotto])(palo_verticale),
                   T([1,3])([x_anta+x_vetro+x*0.05,z_palo_sotto])(palo_verticale),
                   T([1,3])([x_anta+x_vetro+x_palo_verticale,z_palo_sotto])(vetro),
                   T([1,3])([x_anta+x_vetro+x_palo_verticale+x_vetro,z_palo_sotto])(ante),
                   T(3)([z_palo_sotto+z_anta])(palo_sopra)])
    
    #creo il contorno
    if x>4:
        x_contorno=x*1.15
    else:
        x_contorno=x*1.5
    y_contorno=y*0.9
    z_door=SIZE(3)(door)
    y_door=SIZE(2)(door)
    x_pietra=(x_contorno-SIZE(1)(door))/2
    contorno=CUBOID([x_contorno,y_contorno,z_door+x_pietra])
    contorno=DIFFERENCE([contorno,T(1)((SIZE(1)(contorno)-SIZE(1)(door))/2)(S(2)(y*10)(door))])
    contorno=STRUCT([door,T([1,2])([-(SIZE(1)(contorno)-SIZE(1)(door))/2,y_door-y_contorno])(contorno)])
    contorno=COLOR(cPietra)(contorno)

    #assemblo la porta
    door=STRUCT([door,contorno])
    
    if tip==1:
        #aggingo il tetto piccolo sopra alla finestra utilizzando il metodo pilastro_finestra_sopra
        z_sopra=z*0.15   
        x_door=SIZE(1)(door)
        x_sopra=x*1.6
        sopra=COLOR(cPietra)(CUBOID([x_sopra,y_contorno*1.1,z_sopra]))
        door=STRUCT([door,T([1,2,3])([-x_pietra-(x_sopra-x_door)/2,y_door-y_contorno,SIZE(3)(door)])(sopra)])
        x_pilastro=x*2
        pilastro=pilastro_finestra_sopra(x_pilastro,y_contorno*0.5,z_sopra/2)
        pilastro=COLOR(cPietra)(T([1,2,3])([-x_pietra-(x_pilastro-x_door)/2,y_contorno*0.7+y_door-y_contorno,SIZE(3)(door)+z_sopra/2])(pilastro))

        #assemblo
        door=STRUCT([door,pilastro])

    if con_maniglia==1:
        maniglie=maniglia(x/14.0,y/8,z/14.0)
        maniglie_dietro=R([1,2])(PI)(maniglie)
        door=STRUCT([door,
                   T([1,2,3])([x_vetro+x_anta+x*0.02,y,z*1.8])(maniglie),
                   T([1,2,3])([x_vetro+x*0.08+x_anta,y,z*1.8])(maniglie),
                   T([1,3])([x_vetro+x_anta+x*0.02,z*1.8])(maniglie_dietro),
                   T([1,3])([x_vetro+x*0.08+x_anta,z*1.8])(maniglie_dietro)
                  ])
    
    return door

#VIEW(porta(1.3,1,1,1,1))
```

```python
# door crea un porta di legno
# tip==1 con contorno
# tip==0 senza contorno
# con_maniglia==1 con maniglia
# con_maniglia==0 senza maniglia
def door(x,y,z,con_maniglia,tip):
    #creo le varie componenti per la porta
    z_pali=z*0.05
    y=y*0.3
    cLegnoPorta='./image/legno.jpg'
    
    palo_sopra=COLOR(cLegno)(CUBOID([x,y,z_pali]))
    palo_sotto=COLOR(cLegno)(CUBOID([x,y,z_pali]))
    
    x_palo_orizzontale=x*0.455
    z_palo_orizzontale=z*0.05
    palo_orizzontale=COLOR(cLegno)(CUBOID([x_palo_orizzontale,y,z_palo_orizzontale]))
    
    x_palo_verticale=x*0.05
    z_palo_verticale=z*2.95
    palo_verticale=COLOR(cLegno)(mattoncino([x_palo_verticale,y,z_palo_verticale],0))
    
    x_cubo_sopra=x*0.45
    z_cubo_sopra=z*1.15
    cubo_sopra=TEXTURE(cLegnoPorta)(mattoncino([x_cubo_sopra,y,z_cubo_sopra],0))
    
    x_cubo_sotto=x*0.45
    z_cubo_sotto=z*1.7
    cubo_sotto=TEXTURE(cLegnoPorta)(mattoncino([x_cubo_sotto,y,z_cubo_sotto],0))
    
    x_anta=x*0.05
    z_anta=z*3
    ante=COLOR(cLegno)(CUBOID([x_anta,y,z_anta]))
    maniglie=maniglia(x/14.0,y/8,z/14.0)
    
    #unisco il tutto
    porta=STRUCT([ante,T(1)(x_anta)(palo_sotto),
                   T([1,3])([x_anta,z_pali])(cubo_sotto),
                   T([1,3])([x_anta+x_cubo_sotto,z_pali])(palo_verticale),
                   T([1,3])([x_anta+x_cubo_sotto+x_palo_verticale,z_pali])(palo_verticale),
                   T([1,3])([x_anta+x_cubo_sotto+x_palo_verticale*2,z_pali])(cubo_sotto),
                   T(1)(x_anta+x_cubo_sotto+x_palo_verticale*2+x_cubo_sotto)(ante),
                   T([1,3])([x_anta,z_pali+z_cubo_sotto])(palo_orizzontale),
                   T([1,3])([x_anta+x_cubo_sopra+x_palo_verticale*2-x*0.005,z_pali+z_cubo_sotto])(palo_orizzontale),
                   T([1,3])([x_anta,z_pali+z_cubo_sotto+z_palo_orizzontale])(cubo_sopra),
                   T([1,3])([x_anta+x_cubo_sopra+x_palo_verticale*2,z_pali+z_cubo_sotto+z_palo_orizzontale])(cubo_sopra),
                   T([1,3])([x_anta,z_pali+z_cubo_sotto+z_palo_orizzontale+z_cubo_sopra])(palo_sopra)])
    
    if tip==1:
        #creo il contorno
        y_contorno=y*0.9
        z_door=SIZE(3)(porta)
        y_door=SIZE(2)(porta)
        x_pietra=(x*1.5-SIZE(1)(porta))/2
        contorno=CUBOID([x*1.5,y_contorno,z_door+x_pietra])
        contorno=DIFFERENCE([contorno,T([1,2])([(SIZE(1)(contorno)-SIZE(1)(porta))/2,-y])(S(2)(y*10)(porta))])
        contorno=STRUCT([porta,T([1,2])([-(SIZE(1)(contorno)-SIZE(1)(porta))/2,y_door-y_contorno])(contorno)])
        contorno=COLOR(cPietra)(contorno)

        #assemblo la porta
        porta=STRUCT([porta,contorno])

        #aggingo il tetto piccolo sopra alla finestra utilizzando il metodo pilastro_finestra_sopra
        z_sopra=z*0.15   
        x_door=SIZE(1)(porta)
        x_sopra=x*1.6
        sopra=COLOR(cPietra)(CUBOID([x_sopra,y_contorno*1.1,z_sopra]))
        porta=STRUCT([porta,T([1,2,3])([-x_pietra-(x_sopra-x_door)/2,y_door-y_contorno,SIZE(3)(porta)])(sopra)])
        x_pilastro=x*2
        pilastro=pilastro_finestra_sopra(x_pilastro,y_contorno*0.5,z_sopra/2)
        pilastro=COLOR(cPietra)(T([1,2,3])([-x_pietra-(x_pilastro-x_door)/2,y_contorno*0.7+y_door-y_contorno,SIZE(3)(porta)+z_sopra/2])(pilastro))

        #assemblo
        porta=STRUCT([porta,pilastro])
    
    if con_maniglia==1:
        maniglie_dietro=R([1,2])(PI)(maniglie)
        porta=STRUCT([porta,
                    T([1,2,3])([x_anta+x_cubo_sopra+x_palo_verticale-x*0.03,y,z_pali+z_cubo_sotto+z_palo_orizzontale])(maniglie),
                    T([1,2,3])([x_anta+x_cubo_sopra+x_palo_verticale+x*0.03,y,z_pali+z_cubo_sotto+z_palo_orizzontale])(maniglie),
                    T([1,3])([x_anta+x_cubo_sopra+x_palo_verticale-x*0.03,z_pali+z_cubo_sotto+z_palo_orizzontale])(maniglie_dietro),
                    T([1,3])([x_anta+x_cubo_sopra+x_palo_verticale+x*0.03,z_pali+z_cubo_sotto+z_palo_orizzontale])(maniglie_dietro)])

    
    return porta

#VIEW(door(1.3,1,1,1,0))
```

```python
# crea una scalinata
#n numero scalini
def scalini(x,y,z,n):
    y_scalino=y*0.5
    z_scalino=z*1.9
    z_abbassa_scalino=z_scalino/n
    
    
    #i contatore utilizzato anche per abbassare lo scalino
    #y_trasposto di quanto si sposta ogni gradino 
    #scala all'inizio è un booleano per creare il primo scalino e poi diventa un oggetto hpc che rappresenta la scala 
    #il metodo è un metodo ricorsivo che ad ogni passo aggiunge uno scalino 
    def scalini1(i,y_trasposto,scala):
        if i>=n:
            return scala
        else:
            if scala==1:
                y_scalino_1=y_scalino*5
                
                #creo il pavimento dietro piu grande con la parte laterale
                scalino=CUBOID([x*7,y_scalino_1,z_scalino])
                lato=CUBOID([x*0.5,y_scalino_1,z_scalino*1.05])
                scalino=RIGHT([lato,RIGHT([scalino,lato])])
                
                return scalini1(i+1,y_scalino_1,scalino)
            else:
                #creo la parte laterale
                triangolo=MKPOL([[[0,0],[y_scalino,0],[0,z_abbassa_scalino]],[[1,2,3]],[[]]])
                triangolo=PROD([triangolo,QUOTE([x*0.5])])
                triangolo=COMP([R([1,3])(-PI/2),R([1,2])(PI/2)])(triangolo)
                colonna_lato=TOP([CUBOID([x*0.5,y_scalino,(z_scalino-(z_abbassa_scalino*i))*1.05]),triangolo])
                
                #creo lo scalino
                scalino=CUBOID([x*7,y_scalino,z_scalino-(z_abbassa_scalino*i)])
                scalino=RIGHT([colonna_lato,RIGHT([scalino,colonna_lato])])
                
                #assemblo la scala
                scala=STRUCT([scala,T(2)(y_trasposto)(scalino)])
                
                return scalini1(i+1,y_trasposto+y_scalino,scala)
    return scalini1

#VIEW(scalini(1.3,1,1,10)(0,0,1))
```

```python
#questo metodo crea l'arco intorno con due corve di bezier una sopra e una sotto per darne lo spessore
def arco_finestra(x,y,z):
    curva_esterna=BEZIER(S1)([[x*0,z*0],[x*0.5,z*2],[x*2.5,z*4],[x*4.5,z*2],[x*5,z*0]])
    curva_interna=BEZIER(S1)([[x*0.2,z*0],[x*0.5,z*1.8],[x*2.5,z*3.6],[x*4.5,z*1.8],[x*4.8,z*0]]) 
    curva_2D=BEZIER(S2)([curva_esterna,curva_interna]) 
    dominio=PROD([INTERVALS(1)(32),INTERVALS(1)(1)])
    curva_2D=MAP(curva_2D)(dominio)
    return PROD([curva_2D,QUOTE([y*0.5])])

# questo metodo crea la tipologia delle finestre laterali un po curvate fatte con due corve di bezier una verticale per darne lo spessore e l'altra che genera la curva
def finestra_curva_laterale(x,y,z):
    curva_esterna=BEZIER(S1)([[x*0,z*0],[x*0.42,z*1.24],[x*1.4,z*2.08]])
    curva_interna=BEZIER(S1)([[x*1.4,z*0],[x*1.4,z*2.08]])
    curva_2D=BEZIER(S2)([curva_esterna,curva_interna])
    dominio=PROD([INTERVALS(1)(32),INTERVALS(1)(1)])
    curva_2D=MAP(curva_2D)(dominio)
    return PROD([curva_2D,QUOTE([y*0.5])])

# crea la tipologia della finestra al centro fatta con due curve di bezier uno per darni lo spessore e l'altra per la curva in alto
def finestra_centro(x,y,z):
    curva_esterna=BEZIER(S1)([[x*0,z*2.2],[x*0.8,z*2.4],[x*1.6,z*2.2]])
    curva_interna=BEZIER(S1)([[x*0,z*0],[x*1.6,z*0]])
    curva_2D=BEZIER(S2)([curva_esterna,curva_interna])
    dominio=PROD([INTERVALS(1)(32),INTERVALS(1)(1)])
    curva_2D=MAP(curva_2D)(dominio)
    return PROD([curva_2D,QUOTE([y*0.5])])

# questo metodo assembla i metodi sopra descritti e crea la finestra ad arco con l'aggiunta di un asta sotto 
def finestra_arco(x,y,z):
    finestra_curva_laterale_sx=hex_material("#ffffff", "#ffffff", .4)(JOIN(finestra_curva_laterale(x,y*0.95,z)))
    finestra_curva_laterale_dx=COMP([T([1,3])([x*4.9,y*0.485]),R([1,3])(PI)])(finestra_curva_laterale_sx)
    aste=COLOR(cLegno)(CUBOID([x*0.2,z*2.2,y*0.5]))
    finestra_centrale=finestra_centro(x,y*0.95,z)
    finestra_centrale=hex_material("#ffffff", "#ffffff", .4)(JOIN(finestra_centrale))
    arc=COLOR(cLegno)(arco_finestra(x,y,z))
    finestra_senza_asta_sotto=STRUCT([T([1,3])([0.1*x,y*0.013])(finestra_curva_laterale_sx),arc,finestra_curva_laterale_dx,T(1)(x*1.5)(aste),T(1)(x*3.3)(aste),T([1,3])([x*1.7,y*0.013])(finestra_centrale)])
    asta_sotto=COLOR(cLegno)(CUBOID([x*5,z*0.2,y*0.5]))
    return DOWN([finestra_senza_asta_sotto,asta_sotto])

#VIEW(finestra_arco(1,2,1))
```

```python
from larlib import *

#assemblo gli oggetti creati in precedenza su di un muro.
#il muro è stato fatto con due muri: uno sotto e l'altro sopra
#quello sopra è stata aggiunta una sporganza sotto e tra i due muri ho lasciato un buco che poi sarà occupato dal pavimento
def parete_sud(x,y,z):
    
    #creo muro e faccio buchi
    x_wall=x*24
    wall=CUBOID([x_wall,y*0.5,z*8.5])
    windows=finestra(x,y*(0.5/0.3),z,0,1)
    windows_sx=T([1,3])([x*13.5,z*0.7])(finestra(x,y*(0.5/0.3),z,1,1))
    windows_dx=T([1,3])([x*9.5,z*0.7])(finestra(x,y*(0.5/0.3),z,2,1))
    windows=T([1,3])([x*1.5,z*0.7])(windows)
    little_windows=T([1,3])([x*1.5,z*6.5])(finestrella(x,y*(0.5/0.3),z))
    little_windows_dx=STRUCT(NN(2)([little_windows,T(1)(x*4)]))
    little_windows_sx=T(1)(x*16)(little_windows_dx)
    finestra_centrale=finestra_arco(x*(1/1.3),y*2,z)
    door=porta(x*0.97,y*1.6,z,1,0)
    coppia_finestre=STRUCT(NN(2)([windows,T(1)(x*4)]))

    wall=DIFFERENCE([wall,windows_dx,windows_sx,coppia_finestre,T(1)(x*16)(coppia_finestre),little_windows_dx,little_windows_sx,
                     COMP([T([1,2,3])([(x*10.15384),y*0.5*(0.5/0.3),z*5.5]),R([2,3])(PI/2)])(finestra_centrale),
                     T([1,2,3])([x*11.5,-y*0.1,-z*0.2])(door)
                    ])

    #creo gli oggetti per il muro
    windows=finestra(x,y,z,0,1)
    windows_sx=T([1,3])([x*13.5,z*0.7])(finestra(x,y,z,1,1))
    windows_dx=T([1,3])([x*9.5,z*0.7])(finestra(x,y,z,2,1))
    windows=T([1,3])([x*1.5,z*0.7])(windows)
    little_windows=T([1,3])([x*1.5,z*6.5])(finestrella(x,y,z))
    little_windows_dx=STRUCT(NN(2)([little_windows,T(1)(x*4)]))
    little_windows_sx=T(1)(x*16)(little_windows_dx)
    coppia_finestre=STRUCT(NN(2)([windows,T(1)(x*4)]))
    
    #metto gli oggetti nel muro
    wall=STRUCT([wall,coppia_finestre,windows_dx,windows_sx,T(1)(x*16)(coppia_finestre),little_windows_dx,little_windows_sx])
    
    #creo finestra centrale e porta principale
    finestra_centrale=finestra_arco(x*(1/1.3),y,z)
    door=porta(x*0.97,y,z,1,1)

    #assemblo il tutto
    wall=STRUCT([wall,COMP([T([1,2,3])([(x*10.15384),y*0.5,z*5.5]),R([2,3])(PI/2)])(finestra_centrale),
                 T([1,3])([x*11.5,-z*0.2])(door)])
    
    #aggiungo una sporgenza nel muro
    x_sporgenza=((x_wall+0.2)/2)-(x*1.4/2)
    sporgenza=COLOR(GRAY)(CUBOID([x_sporgenza,y*0.6,z*0.2]))
    wall=STRUCT([T([1,3])([x*0.1,z*0.2])(wall),T(1)((x/1.3)*0.05)(S(1)((x/1.3)*0.995)(sporgenza)),
                 T(1)(x_sporgenza+(x*1.43))(sporgenza)])
    
    #faccio muro sotto e buchi
    muro_sotto=CUBOID([x_wall,y*0.5,z*1.8])
    muro_sotto=STRUCT([muro_sotto,T(3)(-z*1.8)(CUBOID([x_wall,y*0.5,z*1.8]))])
    finestra_sotto=finestrella_sotto(x,y*(0.5/0.3),z,0,0)
    finestra_sotto_1=finestrella_sotto(x,y*(0.6/0.3),z,1,0)
    muro_sotto=DIFFERENCE([muro_sotto,T([1,3])([x*5.5,z*0.3])(finestra_sotto),T([1,3])([x*17.5,z*0.3])(finestra_sotto),
                           T([1,2])([x*1.5,-y*0.2])(finestra_sotto_1),T([1,2])([x*21.5,-y*0.2])(finestra_sotto_1)])
    
    #creo gli oggetti del muro sotto
    finestra_sotto=finestrella_sotto(x,y,z,0,0)
    finestra_sotto_1=finestrella_sotto(x,y,z,1,1)
    scale=COLOR(cPietra)(scalini(x,y,z,10)(0,0,1))
    
    #creo i bordi agli angoli
    z_mattone=z*1.8/5.0
    completa=T([1,3])([-z_mattone*0.28,z_mattone*0.1])(COLOR(cMattone)(CUBOID([x*0.1,y*0.1,z_mattone*0.8])))
    mattone_piccolo=COLOR(cMattone)(mattoncino([x*0.25,y*0.5,z_mattone],1))
    mattone_grande=COLOR(cMattone)(mattoncino([x*0.5,y*0.5,z_mattone],1))
    mattoni=STRUCT([mattone_grande,T(3)(z_mattone)(mattone_piccolo),T(3)(z_mattone*2)(mattone_grande),
                    T(3)(z_mattone*3)(mattone_piccolo),T(3)(z_mattone*4)(mattone_grande)])
    mattoni_sx=STRUCT([mattoni,STRUCT(NN(5)([completa,T(3)(z_mattone)])),COMP([T([1,2])([y*0.4,y*0.05]),R([1,2])(PI/2)])(mattoni)])
    mattoni_dx=STRUCT([mattoni,T([1,2])([x*0.1-z_mattone*0.22,y*0.4])(STRUCT(NN(5)([completa,T(3)(z_mattone)]))),COMP([T([1,2])([-y*0.05,y*0.4]),R([1,2])(-PI/2)])(mattoni)])
    
    #assemblo il tutto
    muro_sotto=STRUCT([muro_sotto,T([1,3])([x*5.5,z*0.3])(finestra_sotto),T([1,3])([x*17.5,z*0.3])(finestra_sotto),
                       T(1)(x*1.5)(finestra_sotto_1),T(1)(x*21.5)(finestra_sotto_1),T([1,2])([x*8,y*0.5])(scale),
                       COMP([T([1,2])([x*24.05,y*0.6]),R([1,2])(PI)])(mattoni_sx),T([1,2])([-x*0.02,y*0.1])(mattoni_dx)])
    
    
    return STRUCT([T(1)(x*0.1)(muro_sotto),T(3)(z*2)(wall)])

#VIEW(parete_sud(1.3,1,1))
```

```python
# Questo metodo crea i muri laterali. La variabile tip indica quale tipologia di muro deve creare, visto che quello ad est ed ovest sono uguali ma cambia la posizione del muro più alto.
# I muri laterali cono composti da finestre create sopra e da due muri di cui uno è piu alto e uno piu basso.
# Il muro sotto è stato creato in modo analogo alla parete sud precedente.
# tip==1 muro ovest
# tip==0 muro est
def muri_laterali(x,y,z,tip):
    #muro ovest
    if tip==0:
        #faccio il muro e i vari buchi per le finestre
        x_wall=x*12
        wall=CUBOID([x_wall,y*0.5,z*8.5])
        x_wall2=x*4
        wall2=CUBOID([x_wall2,y*0.5,z*10])
        wall=STRUCT([wall,T(1)(x_wall)(wall2)])
        windows=finestra(x,y*(0.5/0.3),z,0,1)
        windows_sopra=finestra(x,y*(0.5/0.3),z,3,0)
        windows_sotto=T([1,3])([x*1.5,z*0.7])(windows)
        windows_sopra=T([1,3])([x*13.5,z*6.5])(windows_sopra)
        little_windows=T([1,3])([x*1.5,z*6.5])(finestrella(x,y*(0.5/0.3),z))
        wall=DIFFERENCE([wall,windows_sopra,STRUCT(NN(4)([windows_sotto,T(1)(x*4)])),STRUCT(NN(3)([little_windows,T(1)(x*4)]))])


        #creo gli oggetti del muro
        windows=finestra(x,y,z,0,1)
        windows_sopra=finestra(x,y,z,3,0)
        windows_sotto=T([1,3])([x*1.5,z*0.7])(windows)
        windows_sopra=T([1,3])([x*13.5,z*6.5])(windows_sopra)
        little_windows=T([1,3])([x*1.5,z*6.5])(finestrella(x,y,z))
        
        #assemblo il tutto
        wall=STRUCT([wall,windows_sopra,STRUCT(NN(4)([windows_sotto,T(1)(x*4)])),STRUCT(NN(3)([little_windows,T(1)(x*4)]))])

        #creo sporgenza e la metto nel muro
        x_sporgenza=(x_wall+x_wall2+0.2)
        sporgenza=COLOR(GRAY)(CUBOID([x_sporgenza,y*0.6,z*0.2]))
        wall=STRUCT([T([1,3])([x*0.1,z*0.2])(wall),sporgenza])

        #faccio muro sotto e buchi per le finestre
        muro_sotto=CUBOID([x_wall+x_wall2,y*0.5,z*1.8])
        muro_sotto=STRUCT([muro_sotto,T(3)(-z*1.8)(CUBOID([x*16,y*0.5,z*1.8]))])
        finestra_sotto=T([1,3])([x*1.5,z*0.3])(finestrella_sotto(x,y*(0.5/0.3),z,0,0))
        muro_sotto=DIFFERENCE([muro_sotto,STRUCT(NN(3)([finestra_sotto,T(1)(x*4)]))])

        #metto gli oggetti nel muro sotto
        finestra_sotto=T([1,3])([x*1.5,z*0.3])(finestrella_sotto(x,y,z,0,0))
        muro_sotto=STRUCT([muro_sotto,STRUCT(NN(4)([finestra_sotto,T(1)(x*4)]))])
        wall=STRUCT([T(1)(x*0.1)(muro_sotto),T(3)(z*2)(wall)])
        
    #muro est
    else:
        #faccio il muro e i vari buchi per le finestre
        x_wall=x*12
        wall=CUBOID([x_wall,y*0.5,z*8.5])
        x_wall2=x*4
        wall2=CUBOID([x_wall2,y*0.5,z*10])
        wall=STRUCT([wall2,T(1)(x_wall2)(wall)])
        windows=finestra(x,y*(0.5/0.3),z,0,1)
        windows_sopra=finestra(x,y*(0.5/0.3),z,3,0)
        
        windows_sotto=T([1,3])([x*1.5,z*0.7])(windows)
        windows_sopra=T([1,3])([x*1.5,z*6.5])(windows_sopra)
        little_windows=T([1,3])([x*5.5,z*6.5])(finestrella(x,y*(0.5/0.3),z))
        wall=DIFFERENCE([wall,windows_sopra,STRUCT(NN(4)([windows_sotto,T(1)(x*4)])),STRUCT(NN(3)([little_windows,T(1)(x*4)]))])

        #creo gli oggetti del muro
        windows=finestra(x,y,z,0,1)
        windows_sopra=finestra(x,y,z,3,0)
        windows_sotto=T([1,3])([x*1.5,z*0.7])(windows)
        windows_sopra=T([1,3])([x*1.5,z*6.5])(windows_sopra)
        little_windows=T([1,3])([x*5.5,z*6.5])(finestrella(x,y,z))
        
        #assemblo il tutto
        wall=STRUCT([wall,windows_sopra,STRUCT(NN(4)([windows_sotto,T(1)(x*4)])),STRUCT(NN(3)([little_windows,T(1)(x*4)]))])

        #creo sporgenza e la metto nel muro
        x_sporgenza=(x_wall+x_wall2+0.2)
        sporgenza=COLOR(GRAY)(CUBOID([x_sporgenza,y*0.6,z*0.2]))
        wall=STRUCT([T([1,3])([x*0.1,z*0.2])(wall),sporgenza])

        #faccio muro sotto e buchi
        muro_sotto=CUBOID([x_wall+x_wall2,y*0.5,z*1.8])
        muro_sotto=STRUCT([muro_sotto,T(3)(-z*1.8)(CUBOID([x*16,y*0.5,z*1.8]))])
        finestra_sotto=T([1,3])([x*5.5,z*0.3])(finestrella_sotto(x,y*(0.5/0.3),z,0,0))
        muro_sotto=DIFFERENCE([muro_sotto,STRUCT(NN(3)([finestra_sotto,T(1)(x*4)]))])

        #metto gli oggetti nel muro sotto
        finestra_sotto=T([1,3])([x*1.5,z*0.3])(finestrella_sotto(x,y,z,0,0))
        muro_sotto=STRUCT([muro_sotto,STRUCT(NN(4)([finestra_sotto,T(1)(x*4)]))])
        wall=STRUCT([T(1)(x*0.1)(muro_sotto),T(3)(z*2)(wall)])
    
    return wall

#VIEW(muri_laterali(1.3,1,1,1))
#VIEW(muri_laterali(1.3,1,1,0))
```

```python
# scala crea una scala interna al palazzo
# n numero di scalini
# dim_scalino dimensione dello scalino per fare il pavimento
def scala(x,y,z,n,dim_scalino):
    y_scalino=y
    z_scalino=z*1.0
    z_abbassa_scalino=z_scalino/n
    
    #i contatore. utilizzato anche per abbassare lo scalino
    #y_trasposto di quanto si sposta ogni gradino 
    #scala all'inizio è un booleano per creare il primo scalino e poi diventa un oggetto hpc che rappresenta la scala 
    #il metodo è un metodo ricorsivo che ad ogni passo aggiunge uno scalino 
    def scalini1(i,y_trasposto,scala):
        if i>=n:
            return scala
        else:
            if scala==1:
                
                #creo il pavimento dietro piu grande con la parte laterale
                scalino=CUBOID([x,dim_scalino,z_scalino])
                
                return scalini1(i+1,dim_scalino,scalino)
            else:
                #creo lo scalino
                scalino=CUBOID([x,y_scalino,z_scalino-(z_abbassa_scalino*i)])
                
                #assemblo la scala
                scala=STRUCT([scala,T(2)(y_trasposto)(scalino)])
                
                return scalini1(i+1,y_trasposto+y_scalino,scala)
    return scalini1

#VIEW(scala(1.3,0.5,5,10,0.5*3)(0,0,1))
```

```python
#z_altezza altezza muro
#con_porta_sopra==1 muro con porta sopra
#con_porta_sopra==0 muro senza porta sopra
#tip_porta==1 se porta con finestra
#tip_porta==0 se porta di legno
#porta_fuori==1 porta con contorno come foto in precedenza
#porta_fuori==0 porta senza contorno
#tip_grande==1 tipologia di muro con porta piu grande
#tip_grande==0 senza tipologia di muro con porta piu grande
# Questo metodo crea un muro con una porta
def muro_con_porta(x,y,z,z_altezza,con_porta_sopra,tip_porta,porta_fuori,tip_grande):

    x_muro,z_muro=x*1.5,z_altezza
        
    muro=CUBOID([x_muro,y*0.5,z_muro])
    
    if tip_porta==1:
        doors=porta(x,y,z,0,1)
        z_door=SIZE(3)(doors)-z*0.375
    else:
        doors=door(x*0.91,y*1.77,z,1,0)
        z_door=SIZE(3)(doors)    
    if porta_fuori==1:
        doors=COMP([T([1,2])([x,y*0.5]),R([1,2])(PI)])(door(x*0.91,y*1.77,z,1,1))
        z_door=SIZE(3)(doors)-z*0.525

    
    if con_porta_sopra==1:
        x_muro_porta,z_muro_porta=x*1,z*4.8-z_door
        muro_porta=CUBOID([x_muro_porta,y*0.5,z_muro_porta])
        z_muro_sopra=z_muro-z_door*2-z_muro_porta
        muro_sopra=CUBOID([x_muro_porta,y*0.5,z_muro_sopra])
        muro_porta=STRUCT([muro_porta,T(3)(z_muro_porta)(doors),
                     T(3)(z_door+z_muro_porta)(muro_sopra)])
        
    else:
        x_muro_porta,z_muro_porta=x*1,z_muro-z_door
        muro_porta=CUBOID([x_muro_porta,y*0.5,z_muro_porta])
    if tip_grande==1:
        doors=porta(x,y,z,0,1)
        x_size=SIZE(1)(doors)
        doors=porta(x*0.5,y,z,0,1)
        x_size2=SIZE(1)(doors)
        x_size3=x_size-x_size2
        x_muro,z_muro=(x*1.5)+x_size3/4,z_altezza
        muro=CUBOID([x_muro,y*0.5,z_muro])
        x_muro_porta,z_muro_porta=x_size2,z_muro-z_door
        muro_porta=CUBOID([x_muro_porta,y*0.5,z_muro_porta])
        doors=T(1)(x*0.125)(doors)
        
    wall=STRUCT([muro,T([1,3])([x_muro,z_door])(muro_porta),T(1)(x_muro+x_muro_porta)(muro),T(1)(x_muro)(doors)])
      
    return wall

#VIEW(muro_con_porta(1.3,1,1,1*8.7,0,0,0,0))
```

```python
# definisco i vertici che mi serviranno per creare il tetto
def vertici(x,y,z,add1,add2):
    x_pilastro=x
    y_pilastro=y
    z_pilastro=z*0.4
    x_pilastro2=x_pilastro/2
    z_pilastro2=(x_pilastro2/4.1)/COS(PI/6)*2.2
    V=[[0,0,0],#1 quadrato dx
    [0,y_pilastro,0],#2 quadrato dx
    [0,y_pilastro,z_pilastro],#3 quadrato dx
    [0,0,z_pilastro],#4 quadrato dx
    [-y_pilastro+add1,y_pilastro*2,z_pilastro],#5 triangolo dx
    [x_pilastro,0,0],#6 quadrato sx
    [x_pilastro,y_pilastro,0],#7 quadrato sx
    [x_pilastro,y_pilastro,z_pilastro],#8 quadrato sx
    [x_pilastro,0,z_pilastro],#9 quadrato sx
    [x_pilastro+y_pilastro-add2,y_pilastro*2,z_pilastro],#10 triangolo sx
    [0,0,-z_pilastro],#11 sotto dx
    [0,y_pilastro,-z_pilastro],#12 sotto dx
    [x_pilastro,0,-z_pilastro],#13 sotto sx
    [x_pilastro,y_pilastro,-z_pilastro],#14 sotto sx
    [x_pilastro2,-(x_pilastro2-(y_pilastro)),z_pilastro2+z_pilastro],# 15 tetto
    ]
    return V

# Creo il tetto con i vertici sopra unisco le facce estreme e faccio un join per creare l'oggetto
# il tetto è fatto da un cubo sotto, un cubo con angolo sopra e la copertura del tetto sopra 
# x_wall y_wall sono utilizzati per la grandezza del tetto
def tettoia(x,y,z,x_wall,y_wall):
    
    # in questo pezzo assemblo il tetto per i muri che sono piu alti
    V=vertici(x_wall*1.02,y_wall*1.1,z*0.5,0,0)
    
    #base del tetto
    pilastro_rettangolo=MKPOL([V,[[1,4,3,5,2],[6,9,8,10,7]],[]])
    pilastro_rettangolo_cubo=MKPOL([V,[[1,2,11,12],[6,7,13,14]],[]])
    
    #tetto
    su=MKPOL([V,[[15,10,5]],[]])
    pilastro_rettangolo=COLOR(cTegola)(JOIN(pilastro_rettangolo))
    su=COLOR(cTegola)(su)
    pilastro_rettangolo_cubo=JOIN(pilastro_rettangolo_cubo)
    
    #unisco le componenti
    tetto=STRUCT([pilastro_rettangolo_cubo,pilastro_rettangolo,su])
    tetto_dx=COMP([T([1,2])([y_wall*1.1,-x_wall*1.02+y_wall*1.1]),R([1,2])(PI/2)])(tetto)
    tetto_su=COMP([T([1,2])([x_wall*1.02,-x_wall*1.02+y_wall*1.1*2]),R([1,2])(PI)])(tetto)
    tetto_sx=COMP([T([1,2])([x_wall*1.02-y_wall*1.1,y_wall*1.1]),R([1,2])(-PI/2)])(tetto)
    tetto=STRUCT([tetto,tetto_dx,tetto_su,tetto_sx])
    
    return tetto
#VIEW(tettoia(1.3,1,1,5.2,0.9))
```

```python
#con_tetto==1 con tetto
#con_tetto==0 senza tetto
#con_secondo_piano=1 con secondo piano
#con_secondo_piano=0 senza secondo piano
# Creo tutti i muri interni e il tetto, li collego direttamente ai muri est e ovest
def muro_est(x,y,z,con_tetto,con_secondo_piano):
    
    #creo ripologia di muro principale
    wall=T(3)(z*2.2)(muro_con_porta(x,y,z,z*10.2,0,0,1,0))
    x_wall=SIZE(1)(wall)
    y_wall=SIZE(2)(wall)
    
    #creo tetto, scale e pavimenti del secondo piano
    tetto=tettoia(x,y,z,x_wall,y_wall)
    scale=scala(x*1.2,y*0.5,z*5,12,y*0.5*3)(0,0,1)
    x_scala=SIZE(1)(scale)
    scale=COMP([T(3)(z*2.2),R([1,2])(PI/2)])(scale)
    pavimento=T(3)(z*2.2)(COLOR(GRAY)(CUBOID([x*16,y*5,z*0.2])))
    pavimento2=T(3)(z*2.2)(COLOR(GRAY)(CUBOID([x*16/4.1,y*5,z*0.2])))
    

    # Qua creo i muri interni e li collego su uno dei due muri laterali
    muro_laterale=muri_laterali(x,y,z,1)
    
    wall2=T(3)(z*2.2)(muro_con_porta(x,y,z,z*10.2,1,0,0,0))
    wall2=COMP([T([1,2])([x_wall+x*0.1,-x_wall]),R([1,2])(PI/2)])(wall2)
    
    #creo il muro con la finestra
    finestra_centrale=finestra_arco(x*(1/1.3),y*2,z)
    finestra_centrale=COMP([T([1,3])([(x*8-SIZE(1)(finestra_centrale))/2,z*1]),R([2,3])(PI/2)])(finestra_centrale)
    wall8=CUBOID([x*8,y*0.5,z*3.7])
    wall_finestra=DIFFERENCE([wall8,T(2)(y*0.75)(finestra_centrale)])
    finestra_centrale=finestra_arco(x*(1/1.3),y,z)
    finestra_centrale=COMP([T([1,3])([(x*8-SIZE(1)(finestra_centrale))/2,z*1]),R([2,3])(PI/2)])(finestra_centrale)
    wall_finestra=STRUCT([wall_finestra,T(2)(y*0.5)(finestra_centrale)])
    
    #creo muri con porte
    wall3=muro_con_porta(x,y,z,z*5,0,0,0,0)
    wall3=STRUCT([T(1)(SIZE(1)(wall3))(CUBOID([x*4,y*0.5,z*5])),wall3])
    wall_finestra=STRUCT([wall3,T(3)(z*5)(wall_finestra)])
    wall3=STRUCT([wall3,T(3)(z*5)(wall8)])
    wall3=T(3)(z*2.2)(wall3)
    wall_finestra=T(3)(z*2.2)(wall_finestra)
    x_wall3=SIZE(1)(wall3)
        
    wall4=T(3)(z*2.2)(muro_con_porta(x,y,z,z*8.7,1,0,0,0))
    wall4=COMP([T([1,2])([x_wall+x*0.4+x_wall3,-x_wall]),R([1,2])(PI/2)])(wall4)
        
    wall5=T(3)(z*2.2)(muro_con_porta(x,y,z,z*8.7,0,0,0,0))
    
    x_wall6=x*4
    doors=door(x*0.91,y*3,z,0,0)
    x_door=x
    wall6=T(3)(z*2.2)(CUBOID([x_wall6,y*0.5,z*8.7]))
    wall6=DIFFERENCE([wall6,T([1,2,3])([x_wall6-y*0.1-x_door,-y*0.25,z*7.2])(doors)])
    doors=door(x*0.91,y*1.77,z,1,0)
    wall6=STRUCT([wall6,T([1,3])([x_wall6-y*0.1-x_door,z*7.2])(doors)])
    wall6=COMP([T([1,2])([x_wall+x*0.4+x_wall3,-x_wall*2]),R([1,2])(PI/2)])(wall6)
    
    wall7=T(3)(z*2.2)(muro_con_porta(x*3.9,y,z,z*8.7,0,1,0,1))
    wall7=COMP([T([1,2])([x_wall+x*0.1,-x_wall-SIZE(1)(wall7)]),R([1,2])(PI/2)])(wall7)
         
    #assemblo il tutto
    pareti=STRUCT([T([1,2])([x*0.1,-y*x_wall])(wall),
                    wall2,T([1,2])([x_wall+x*0.1,-y*x_wall])(wall3),
                    T([1,2])([x_wall+x*0.1+SIZE(1)(scale)-SIZE(1)(scale)+x_wall3,-y*x_wall-SIZE(2)(scale)])(scale),
                    T([1,2])([x_wall+x*0.1,-y*x_wall-x_scala-y*0.5])(wall_finestra),wall4,
                    T([1,2])([x_wall+x*0.1+x_wall3,-y*x_wall-x_wall6])(wall5),
                    wall6,wall7])
    pareti=STRUCT([muro_laterale,T(3)(-z*0.2)(pareti)])

    #aggiungo il secondo piano
    if con_secondo_piano==1:

        pareti=STRUCT([pareti,T([1,2,3])([x*0.1,-SIZE(2)(pavimento),z*4.6])(pavimento),
                    T([1,2,3])([x*0.1+SIZE(1)(pavimento)-SIZE(1)(pavimento2),-SIZE(2)(pavimento)*2,z*4.6])(pavimento2)])

    #creo ed aggiungo il tetto piu grande
    if con_tetto==1:
        #qua creo il tetto per la parte centrale della casa collegato allo stesso muro laterale degli interni
        #prendo i vertici creati nel metodo in precedenza facendo una parte del tetto
        V=vertici((x_wall*3)+SIZE(2)(wall),y_wall*1.1,z*0.5,0.6,0)
        
        #collego i vari vertici
        pilastro_rettangolo=MKPOL([V,[[1,4,3,5,2],[6,9,8,10,7]],[]])
        pilastro_rettangolo_cubo=MKPOL([V,[[1,2,11,12],[6,7,13,14]],[]])
        su=MKPOL([V,[[15,10,5]],[]])
            
        #unisco i pezzi del tetto
        pilastro_rettangolo=COLOR(cTegola)(JOIN(pilastro_rettangolo))
        su=COLOR(cTegola)(su)
        pilastro_rettangolo_cubo=JOIN(pilastro_rettangolo_cubo)
            
        #metto i vari pezzi nelle posizioni corrette
        x_sposta_tetto,z_sposta_tetto=x_wall3/2,SIZE(3)(wall3)+z*2.2+SIZE(3)(pilastro_rettangolo_cubo)
        pilastro_rettangolo_cubo=T([1,3])([x_sposta_tetto,z_sposta_tetto])(pilastro_rettangolo_cubo)
        pilastro_rettangolo=T([1,3])([x_sposta_tetto,z_sposta_tetto])(pilastro_rettangolo)
        su=T([1,3])([x_sposta_tetto,z_sposta_tetto])(su)
        
        #creo l'altra parte del tetto facendo operazioni analoge a prima
        V2=vertici((x_wall*3)+SIZE(2)(wall),y_wall*1.1,z*0.5,0,0.6)
        
        #collego i vari vertici
        pilastro_rettangolo_opposto=MKPOL([V2,[[1,4,3,5,2],[6,9,8,10,7]],[]])
        pilastro_rettangolo_cubo_opposto=MKPOL([V2,[[1,2,11,12],[6,7,13,14]],[]])
        su_opposto=MKPOL([V2,[[15,10,5]],[]])
            
        #unisco i pezzi del tetto
        pilastro_rettangolo_opposto=COLOR(cTegola)(JOIN(pilastro_rettangolo_opposto))
        su_opposto=COLOR(cTegola)(su_opposto)
        pilastro_rettangolo_cubo_opposto=JOIN(pilastro_rettangolo_cubo_opposto)
        
        #metto i vari pezzi nelle posizioni corrette
        x_sposta_tetto_opposto,z_sposta_tetto_opposto=x_wall3/2,SIZE(3)(wall3)+z*2.2+SIZE(3)(pilastro_rettangolo_cubo_opposto)
        pilastro_rettangolo_cubo_opposto=T([1,3])([x_sposta_tetto_opposto,z_sposta_tetto_opposto])(pilastro_rettangolo_cubo_opposto)
        pilastro_rettangolo_opposto=T([1,3])([x_sposta_tetto_opposto,z_sposta_tetto_opposto])(pilastro_rettangolo_opposto)
        su_opposto=T([1,3])([x_sposta_tetto_opposto,z_sposta_tetto_opposto])(su_opposto)
        
        x_sposta,y_sposta=SIZE(1)(pilastro_rettangolo_cubo)+SIZE(1)(wall)*2,-SIZE(1)(wall)*6+SIZE(2)(pilastro_rettangolo_cubo)+SIZE(2)(pilastro_rettangolo_cubo)/2
        pilastro_rettangolo_cubo_opposto=COMP([T([1,2])([x_sposta,y_sposta]),R([1,2])(PI)])(pilastro_rettangolo_cubo_opposto)
        pilastro_rettangolo_opposto=COMP([T([1,2])([x_sposta,y_sposta]),R([1,2])(PI)])(pilastro_rettangolo_opposto)
        su_opposto=COMP([T([1,2])([x_sposta,y_sposta]),R([1,2])(PI)])(su_opposto)
            
        #parte in basso del tetto che crea la sporgenza rettangolare
        pilastro_rettangolo_cubo=JOIN([pilastro_rettangolo_cubo,pilastro_rettangolo_cubo_opposto])
        #parte centrale del tetto che crea le sporgenze triangolari
        pilastro_rettangolo=COLOR(cTegola)(JOIN([pilastro_rettangolo_opposto,pilastro_rettangolo]))
        #parte sopra del tetto
        su=COLOR(cTegola)(JOIN([su,su_opposto]))
        
        #assemlbo il tutto
        tetto_grande=STRUCT([pilastro_rettangolo_cubo,pilastro_rettangolo,su])
            
        x_spostamento=(x*0.08+x_wall+x_wall3)-(((x_wall*1.02)-x_wall)/2)
        y_spostamento=y_wall*1.1
        z_spostamento=z*14+z*0.5*0.4-z*1.8             
        pareti=STRUCT([pareti,T([2,3])([-y*0.2,-z*0.2])(tetto_grande),T([2,3])([-y*0.3,z_spostamento])(S([1,2])([(x/1.3)*1.04,(x_wall+y_wall)/x_wall*1.02])(tetto))])

    return pareti
        
             
#VIEW(muro_est(1.3,1,1,1,0))
#VIEW(muro_est(1.3,1,1,0,0))
#VIEW(muro_est(1.3,1,1,0,1))
```

```python
#spessore_grande e piccolo da un arco a l'altro
#grandezza_contorno la grandezza del bordo da un mattone ad un altro in percentuale, quindi un numero che va da 0 ad 1
#numero_mattoni numero di mattoni nell'arco
#y spessore della profondità 3d
def arco(x,y,z,spessore_grande,spessore_piccolo,numero_mattoni):
    grandezza_mattone=180.0/numero_mattoni
    grandezza_contorno=grandezza_mattone*0.1
    
    #cont parte da 1 ed è il contatore in base al numero di mattoni
    #struttra all'inizio coniente un intero per il primo elemento dell'oggetto hpc, dopo diventa l'oggetto hpc che contiene l'arco
    #il tutto è stato fatto con una funzione ricorsiva che computa un numero di volte in base al numero di mattoni
    def arco1(cont,struttura):
        #passo iniziale
        if struttura==1:            
            struttura=MKPOL(([[x*spessore_piccolo*COS(radians(grandezza_mattone*(cont-1))),z*spessore_piccolo*SIN(radians(grandezza_mattone*(cont-1)))],
                              [x*spessore_grande*COS(radians(grandezza_mattone*(cont-1))),z*spessore_grande*SIN(radians(grandezza_mattone*(cont-1)))],
                              [x*spessore_piccolo*COS(radians(grandezza_mattone*cont-grandezza_contorno)),z*spessore_piccolo*SIN(radians(grandezza_mattone*cont-grandezza_contorno))],
                              [x*spessore_grande*COS(radians(grandezza_mattone*cont-grandezza_contorno)),z*spessore_grande*SIN(radians(grandezza_mattone*cont-grandezza_contorno))]],
                             [[1,2,4,3]],[]))
            
            contorno=MKPOL(([[x*spessore_piccolo*COS(radians(grandezza_mattone*cont-grandezza_contorno)),z*spessore_piccolo*SIN(radians(grandezza_mattone*cont-grandezza_contorno))],
                             [x*spessore_grande*COS(radians(grandezza_mattone*cont-grandezza_contorno)),z*spessore_grande*SIN(radians(grandezza_mattone*cont-grandezza_contorno))],
                             [x*spessore_piccolo*COS(radians(grandezza_mattone*cont)),z*spessore_piccolo*SIN(radians(grandezza_mattone*cont))],
                             [x*spessore_grande*COS(radians(grandezza_mattone*cont)),z*spessore_grande*SIN(radians(grandezza_mattone*cont))]],
                            [[1,2,4,3]],[]))
            
            struttura=PROD([struttura,QUOTE([y])])
            contorno=PROD([contorno,QUOTE([y*0.9])])
            struttura=STRUCT([struttura,contorno])
            return arco1(cont+1,struttura)
        else:
            #passo base
            if cont==numero_mattoni:
                struttura2=MKPOL(([[x*spessore_piccolo*COS(radians(grandezza_mattone*(cont-1))),z*spessore_piccolo*SIN(radians(grandezza_mattone*(cont-1)))],
                              [x*spessore_grande*COS(radians(grandezza_mattone*(cont-1))),z*spessore_grande*SIN(radians(grandezza_mattone*(cont-1)))],
                              [x*spessore_piccolo*COS(radians(grandezza_mattone*cont)),z*spessore_piccolo*SIN(radians(grandezza_mattone*cont))],
                              [x*spessore_grande*COS(radians(grandezza_mattone*cont)),z*spessore_grande*SIN(radians(grandezza_mattone*cont))]],
                             [[1,2,4,3]],[]))
        
                
                struttura2=PROD([struttura2,QUOTE([y])])
                struttura=STRUCT([struttura,struttura2])
                return COLOR(cMattone)(struttura)
            #passo induttivo
            else:
                struttura2=MKPOL(([[x*spessore_piccolo*COS(radians(grandezza_mattone*(cont-1))),z*spessore_piccolo*SIN(radians(grandezza_mattone*(cont-1)))],
                              [x*spessore_grande*COS(radians(grandezza_mattone*(cont-1))),z*spessore_grande*SIN(radians(grandezza_mattone*(cont-1)))],
                              [x*spessore_piccolo*COS(radians(grandezza_mattone*cont-grandezza_contorno)),z*spessore_piccolo*SIN(radians(grandezza_mattone*cont-grandezza_contorno))],
                              [x*spessore_grande*COS(radians(grandezza_mattone*cont-grandezza_contorno)),z*spessore_grande*SIN(radians(grandezza_mattone*cont-grandezza_contorno))]],
                             [[1,2,4,3]],[]))
            
                contorno2=MKPOL(([[x*spessore_piccolo*COS(radians(grandezza_mattone*cont-grandezza_contorno)),z*spessore_piccolo*SIN(radians(grandezza_mattone*cont-grandezza_contorno))],
                             [x*spessore_grande*COS(radians(grandezza_mattone*cont-grandezza_contorno)),z*spessore_grande*SIN(radians(grandezza_mattone*cont-grandezza_contorno))],
                             [x*spessore_piccolo*COS(radians(grandezza_mattone*cont)),z*spessore_piccolo*SIN(radians(grandezza_mattone*cont))],
                             [x*spessore_grande*COS(radians(grandezza_mattone*cont)),z*spessore_grande*SIN(radians(grandezza_mattone*cont))]],
                            [[1,2,4,3]],[]))
                
                struttura2=PROD([struttura2,QUOTE([y])])
                contorno2=PROD([contorno2,QUOTE([y*0.9])])
                struttura2=STRUCT([struttura2,contorno2])
                struttura=STRUCT([struttura,struttura2])
                return arco1(cont+1,struttura)
                    
    return arco1

#VIEW(arco(1.3,1,1,1.85,0.85,13)(1,1))
```

```python
# Creo la colonnina che sta sotto gli archi. Fatta con 4 curve di bezier con punti fatti come una serpentina che poi ho collegato
def colonnina(x,y,z):
    x=x/1.3
    curva_sx=BEZIER(S1)([[x*1,y*0,z*0],[-x*2,y*0,z*1],[x*5,y*0,z*2],[-x*2,y*0,z*3],[x*1,y*0,z*4]])
    curva_up=BEZIER(S1)([[x*0,-y*1,z*0],[x*0,y*2,z*1],[x*0,-y*5,z*2],[x*0,y*2,z*3],[x*0,-y*1,z*4]]) 
    curva_2D=BEZIER(S2)([curva_sx,curva_up]) 
  
    curva_dx=BEZIER(S1)([[-x*1,y*0,z*0],[x*2,y*0,z*1],[-x*5,y*0,z*2],[x*2,y*0,z*3],[-x*1,y*0,z*4]]) 
    curva_down=BEZIER(S1)([[x*0,y*1,z*0],[x*0,-y*2,z*1],[x*0,y*5,z*2],[x*0,-y*2,z*3],[x*0,y*1,z*4]])
    curva_2D1=BEZIER(S2)([curva_dx,curva_down]) 
    
    dominio1=PROD([INTERVALS(1)(32),INTERVALS(1)(1)])

    curva_3D=BEZIER(S3)([curva_2D,curva_2D1]) 
    dominio3d=PROD([dominio1,QUOTE([1])])
    curva_3D=MAP(curva_3D)(dominio3d)
    return COLOR(cMattone)(curva_3D)

#VIEW(colonnina(1.3,1,1))
```

```python
# Creo il muretto con le colonnine create sopra ci sono due aste sopra e sotto e un cubo al centro
def muretto(x,y,z):
    
    col=R([1,2])(PI/4)(colonnina(x,y,z))
    x_dim_col,y_dim_col=SIZE(1)(col),SIZE(2)(col)
    muretto=STRUCT(NN(5)([col,T(1)((x_dim_col)+(x_dim_col)/10)]))
    mur_centro=T([1,2])([-x_dim_col/2,-y_dim_col/2])(CUBOID([x_dim_col,y_dim_col,SIZE(3)(col)]))
    muretto=STRUCT([muretto,T(1)(SIZE(1)(muretto)+(x_dim_col)/10)(mur_centro),
                   T(1)(SIZE(1)(mur_centro)+SIZE(1)(muretto)+(x_dim_col)*2/10)(muretto)])
    x_asta=SIZE(1)(muretto)*1.1
    z_asta=SIZE(3)(muretto)/10
    asta=CUBOID([x_asta,SIZE(2)(muretto),z_asta])
    muretto=T([1,2,3])([(x_dim_col/2)+(x_asta-(x_asta/1.1))/2,y_dim_col/2,z_asta])(TOP([BOTTOM([muretto,asta]),asta]))
    
    return COLOR(cMattone)(muretto)

#VIEW(muretto(1.3,1,1))
```

```python
#cornice che andrà sopra gli archi
#num è il numero di colonnine nella cornice
def cornice(arg,num):
    x,y,z=arg
    x=x/num
    sfondo=CUBOID([x*num,y*0.8,z*0.8])
    sporgenza=CUBOID([x*num,y,z*0.1])
    mattone=mattoncino([x*0.1,y*0.7,z*0.8],2)
    colonnine=T([1,2])([x*0.35,y*0.2])(STRUCT(NN(3)([mattone,T(1)(x*0.1)])))
    temp=STRUCT([sfondo,STRUCT(NN(num)([colonnine,T(1)(x)]))])
    return COLOR(cMattone)(STRUCT([sporgenza,T(3)(z*0.1)(temp),T(3)(z*0.9)(sporgenza)]))

#VIEW(cornice([7.0,0.52,0.5],9))
```

```python
# creo gli archi con due colonne accoppiate a sinistra e a destra, una colonna centrale e tre archi di cui due hanno il muretto creato sopra
def archi(x,y,z):
    depth=y*0.5
    x_mattone=x*0.8
    z_mattone=z*0.3625
    n=11
    
    #creo arco
    mattone=COLOR(cMattone)(mattoncino([x_mattone,depth,z_mattone],0))
    colonna=STRUCT(NN(n)([mattone,T(3)(z*0.3625)]))
    lunghezza_arco=1.3*0.75*3.75
    w=1.3*0.75
    arch=arco(x,depth,z,1.85,0.85,13)(1,1)
    arch=STRUCT([T(1)(-(1.3/(lunghezza_arco+w)))(colonna),T(1)(lunghezza_arco+(1.3/(lunghezza_arco+w))-x_mattone)(colonna),
                 COMP([T([1,3])([x*1.41,z_mattone*n]),R([1,2])(PI),R([2,3])(PI/2)])(arch)])
    
    #creo colonna
    x_mattone2=x*0.8
    y_mattone2=y*0.52
    mattone2=COLOR(cMattone)(mattoncino([x_mattone2,y_mattone2,z_mattone],0))
    x_mattone_sopra=x_mattone2/2
    mattone_sopra=COLOR(cMattone)(mattoncino([x_mattone_sopra,y_mattone2,z_mattone],0))
    colonna=STRUCT(NN(16)([mattone2,T(3)(z_mattone)]))
    colonna_centro=STRUCT([T([1,3])([x_mattone2,z_mattone*15])(mattone_sopra),T([1,3])([x_mattone2*1.5,z_mattone*15])(mattone_sopra),colonna,
                           T([1,3])([-x_mattone_sopra,z_mattone*15])(mattone_sopra),T([1,3])([-x_mattone_sopra*2,z_mattone*15])(mattone_sopra)])
    
    #assemblo con le colonne arco destra
    colonna_sx=STRUCT([T([1,3])([x_mattone2,z_mattone*15])(mattone_sopra),T([1,3])([x_mattone2*1.5,z_mattone*15])(mattone_sopra),colonna])
    colonna_dx=RIGHT([colonna,colonna_sx])
    arco_dx=DIFFERENCE([T(1)(x_mattone2+x_mattone2*0.5)(arch),S(2)(y*1.5)(T(2)(-y*0.1)(colonna_dx)),
                       S(2)(y*1.5)(T([1,2])([(x/1.3)*lunghezza_arco+x_mattone2,y*0.1])(colonna_centro))])
    arco_dx=COLOR(cMattone)(STRUCT([arco_dx,T(2)(z*-0.01)(colonna_dx)]))
    
    #assemblo con le colonne arco centro
    arco_centro=DIFFERENCE([T(1)(x_mattone2*0.5)(arch),S(2)(y*1.5)(T(2)(-y*0.1)(colonna_centro)),
                        S(2)(y*1.5)(T([1,2])([(x/1.3)*lunghezza_arco,-y*0.1])(colonna_centro))])
    arco_centro=COLOR(cMattone)(STRUCT([arco_centro,T(2)(z*-0.01)(colonna_centro),
                        T([1,2])([(x/1.3)*lunghezza_arco,z*-0.01])(colonna_centro)]))
    
    #assemblo con le colonne arco sinistra
    colonna_sxx=STRUCT([T([1,3])([-x_mattone_sopra,z_mattone*15])(mattone_sopra),T([1,3])([-x_mattone_sopra*2,z_mattone*15])(mattone_sopra),colonna])
    colonna_dxx=LEFT([colonna,colonna_sxx])
    arco_sx=DIFFERENCE([arch,S(2)(y*1.5)(T([1,2])([(x/1.3)*lunghezza_arco+x_mattone2*0.5,y*0.1])(colonna_dxx)),
                       S(2)(y*1.5)(T([1,2])([-x*0.4,y*0.1])(colonna_centro))])
    arco_sx=COLOR(cMattone)(STRUCT([arco_sx,T([1,2])([(x/1.3)*lunghezza_arco+x_mattone2*0.5,z*-0.01])(colonna_dxx)]))
    
    #unisco il tutto
    archi=STRUCT([arco_dx,T(1)((x/1.3)*lunghezza_arco+x_mattone2)(arco_centro),
                                   T(1)((x/1.3)*lunghezza_arco*2+x_mattone2+x_mattone2*0.5)(arco_sx)])
    asta_sopra=cornice([SIZE(1)(archi),y_mattone2,z*0.5],22)
    archi=STRUCT([archi,T([2,3])([y*0.05,z_mattone*16])(COLOR(cMattone)(asta_sopra))])
    
    x1=x/1.3
    #aggiungo il muretto
    muro_colonnine=S([1,2,3])([x1*0.12,y*0.36,z*0.2])(muretto(x,y,z))
    
    archi=STRUCT([archi,T(1)(x_mattone2*2+x_mattone2*0.18)(muro_colonnine),
                  T(1)(SIZE(1)(archi)-(x_mattone2*2+x_mattone2*0.18)-SIZE(1)(muro_colonnine))(muro_colonnine)])
    archi=S([1,3])([(x/1.2975)*1.41,z*0.9])(archi)
    return archi

#VIEW(archi(1.3,1,1))
```

```python
#creo pilastro tetto che andrà sopra aglia archi. E' stato fatto con un mkpol. Ho creato le varie facce e poi ho creato l'oggetto con vari join
#con_tetto==1 con tetto
#con_tetto==0 senza tetto
def pilastro_tetto(x,y,z,con_tetto):
    
    #definisco i punti
    x_pilastro=x*1.01
    y_pilastro=y*1.05
    z_pilastro=z*0.4
    x_pilastro2=x_pilastro/2
    z_pilastro2=(x_pilastro2/4.1)/COS(PI/6)
    
    V=[[0,0,0],#1 quadrato dx
    [0,y_pilastro,0],#2 quadrato dx
    [0,y_pilastro,z_pilastro],#3 quadrato dx
    [0,0,z_pilastro],#4 quadrato dx
    [0,y_pilastro*2,z_pilastro],#5 triangolo dx
    [x_pilastro,0,0],#6 quadrato sx
    [x_pilastro,y_pilastro,0],#7 quadrato sx
    [x_pilastro,y_pilastro,z_pilastro],#8 quadrato sx
    [x_pilastro,0,z_pilastro],#9 quadrato sx
    [x_pilastro,y_pilastro*2,z_pilastro],#10 triangolo sx
    [x_pilastro2,0,z_pilastro2],#11 quadrato su
    [x_pilastro2,y_pilastro,z_pilastro2],#12 quadrato su
    [x_pilastro2,y_pilastro,z_pilastro2+z_pilastro],#13 quadrato su
    [x_pilastro2,0,z_pilastro2+z_pilastro],#14 quadrato su
    [x_pilastro2,y_pilastro*2,z_pilastro2+z_pilastro],#15 triangolo su
    [0,0,-z_pilastro],#16 sotto dx
    [0,y_pilastro,-z_pilastro],#17 sotto dx
    [x_pilastro,0,-z_pilastro],#18 sotto sx
    [x_pilastro,y_pilastro,-z_pilastro],#19 sotto sx
    [x_pilastro2,0,z_pilastro2-z_pilastro], #20 sotto su
    [x_pilastro2,y_pilastro,z_pilastro2-z_pilastro], #21 sotto su
    [x_pilastro2,-(x/(1.3*3.6)),z_pilastro2-(z_pilastro*1.5)],# 22 tetto
    [y_pilastro,-(x/(1.3*3.6)),z_pilastro2-(z_pilastro*1.5)],# 23 tetto
    [x_pilastro-y_pilastro,-(x/(1.3*3.6)),z_pilastro2-(z_pilastro*1.5)],# 24 tetto
    [y_pilastro,0,z_pilastro],# 25 tetto
    [x_pilastro-y_pilastro,0,z_pilastro]# 26 tetto
    ]
    
    #collego i punti
    pilastro_angolo_sx=MKPOL([V,[ [6,9,8,10,7],[11,14,13,15,12]],[]])    
    pilastro_angolo_sx_cubo=MKPOL([V,[[6,7,18,19],[11,12,20,21]],[]])
    pilastro_angolo_dx=MKPOL([V,[[1,4,3,5,2],[11,14,13,15,12]],[]])
    pilastro_angolo_dx_cubo=MKPOL([V,[[1,2,16,17],[11,12,20,21]],[]])
    pilastro_rettangolo=MKPOL([V,[[1,4,3,5,2],[6,9,8,10,7]],[]])
    pilastro_rettangolo_cubo=MKPOL([V,[[1,2,16,17],[6,7,18,19]],[]])
    sfondo=MKPOL([V,[[4,9,11]],[]])
    
    pilastro_angolo_sx=JOIN(pilastro_angolo_sx)
    pilastro_angolo_sx_cubo=JOIN(pilastro_angolo_sx_cubo)
    pilastro_angolo_dx=JOIN(pilastro_angolo_dx)
    pilastro_angolo_dx_cubo=JOIN(pilastro_angolo_dx_cubo)
    pilastro_rettangolo=JOIN(pilastro_rettangolo)
    pilastro_rettangolo_cubo=JOIN(pilastro_rettangolo_cubo)
    sfondo=COMP([T(1)(x_pilastro),R([1,2])(PI)])(TEXTURE('./image/stemma.png')(JOIN(sfondo)))
    pilastri=STRUCT([sfondo,pilastro_angolo_sx,pilastro_angolo_sx_cubo,pilastro_angolo_dx,
            pilastro_angolo_dx_cubo,pilastro_rettangolo,pilastro_rettangolo_cubo])
    # creo il tetto
    if con_tetto==1:
        pilastro_angolo_sx=COLOR(cTegola)(pilastro_angolo_sx)
        pilastro_angolo_dx=COLOR(cTegola)(pilastro_angolo_dx)
        tetto=COLOR(cTegola)(STRUCT([JOIN(MKPOL([V,[[14,22,26]],[]])),MKPOL([V,[[14,25,22]],[]]),
                                     JOIN(MKPOL([V,[[22,23,25]],[]])),JOIN(MKPOL([V,[[26,24,22]],[]]))]))
        tetto=STRUCT([tetto,pilastro_angolo_sx,pilastro_angolo_sx_cubo,pilastro_angolo_dx,pilastro_angolo_dx_cubo])
        tetto=T(3)(SIZE(3)(pilastro_rettangolo)+SIZE(3)(pilastro_rettangolo_cubo))(tetto)
        pilastri=STRUCT([pilastri,tetto])
    
    return pilastri

#VIEW(pilastro_tetto(20.9364624023,0.519999980927,1,1))
```

```python
# Metodo che mi crea una semi circonferenza
def scala_cilidriche(x,y,z):

    x1= lambda (u,v) : v*cos(u) 
    y1= lambda (u,v) : v*sin(u)

    dom= PROD([INTERVALS(PI)(32),INTERVALS(1)(32)])
    mapp= MAP([x1,y1])(dom)
    
    return JOIN(S([1,2])([(x/1.3),y])(PROD([mapp,QUOTE([z*2.0/6.0])])))

#Metodo ricorsivo che crea gli scalini della scala cilidrica. Ad ogni passo aggiunge uno scalino sopra
def scale_cilindriche(x,y,z):
    # n numero scalini
    # scala prima è 1 per il primo scalino poi diventa un oggetto hpc scala
    # i serve per mettere gli scalini uno sopra l'altro
    def scalini(n,scala,i):
        if n<1 :
            return scala
        else:
            scalino=scala_cilidriche(x*(n+2),y*(n+2),z)
            if scala==1:   
                return scalini(n-1,scalino,i+2.0/6.0)
            else:
                scala=STRUCT([scala,T(3)(i)(scalino)])
                return scalini(n-1,scala,i+2.0/6.0)
    return scalini
                

#VIEW(scale_cilindriche(1.3,1,1)(6,1,0))
```

```python
# Metodo che crea la parete nord, assembla i pezzi creati sopra e la tipologia di muro sotto e sopra è
#con_tetto==1 con tetto
#con_tetto==0 senza tetto
def parete_nord(x,y,z,con_tetto):  
    #faccio muro
    x_wall=x*4
    wall=CUBOID([x_wall,y*0.5,z*10])
    windows=finestra(x,y*(0.5/0.3),z,0,1)
    windows_sopra=T([1,3])([x*1.5,z*0.7])(finestra(x,y*(0.5/0.3),z,3,0))
    windows=T([1,3])([x*1.5,z*0.7])(windows)
    wall=DIFFERENCE([wall,windows,T(3)(z*5.8)(windows_sopra)])

    #creo gli oggetti del muro
    windows=finestra(x,y,z,0,1)
    windows=T([1,3])([x*1.5,z*0.7])(windows)
    windows_sopra=T([1,3])([x*1.5,z*0.7])(finestra(x,y,z,3,0))
    wall=STRUCT([wall,windows,T(3)(z*5.8)(windows_sopra)])
    x_sporgenza=(x_wall+0.2)
    sporgenza=COLOR(GRAY)(CUBOID([x_sporgenza,y*0.6,z*0.2]))
    
    #assemblo il tutto
    wall=STRUCT([T([1,3])([x*0.1,z*0.2])(wall),sporgenza])
    
    #faccio muro sotto e buchi
    z_muro_sotto=z*1.8
    muro_sotto=CUBOID([x_wall,y*0.5,z_muro_sotto])
    finestra_sotto=T([1,3])([x*1.5,z*0.3])(finestrella_sotto(x,y*(0.5/0.3),z,0,0))
    muro_sotto=DIFFERENCE([muro_sotto,finestra_sotto])
    
    #metto gli oggetti nel muro sotto
    finestra_sotto=T([1,3])([x*1.5,z*0.3])(finestrella_sotto(x,y,z,0,0))
    muro_sotto=STRUCT([muro_sotto,finestra_sotto])
    wall=STRUCT([T(1)(x*0.1)(muro_sotto),T(3)(z_muro_sotto+(z*0.2))(wall)])
    
    #creo gli archi
    arch=S(1)((x/1.3)*1.05)(archi(x,y,z)) 

    #creo il tetto
    pilastro=pilastro_tetto(20.9364624023,0.519999980927,z,con_tetto)
    arco=STRUCT([S(2)(y*1.1)(arch),T([1,3])([(-SIZE(1)(arch)*(1.005))+SIZE(1)(arch),SIZE(3)(arch)+z*0.4])(pilastro)])
    
    #creo muro sotto agli archi
    sotto_archi=COLOR(cMattone)(mattonata([SIZE(1)(arco),y*0.5,z_muro_sotto],10,4))
    arco=STRUCT([sotto_archi,T(3)(z_muro_sotto+(z*0.2))(arco)])
    #assemblo
    muro_sotto2=T([1,3])([x*0.1,-z*1.8])(CUBOID([x*24,y*0.5,z*1.8]))
    
    #creo la scala
    scale=COLOR(cPietra)(scale_cilindriche(x/1.87,y/1.87,z)(6,1,0))
    
    #creo i bordi agli angoli
    z_mattone=z*1.8/5.0
    completa=T([1,3])([-z_mattone*0.28,z_mattone*0.1])(COLOR(cMattone)(CUBOID([x*0.1,y*0.1,z_mattone*0.8])))
    mattone_piccolo=COLOR(cMattone)(mattoncino([x*0.25,y*0.5,z_mattone],1))
    mattone_grande=COLOR(cMattone)(mattoncino([x*0.5,y*0.5,z_mattone],1))
    mattoni=STRUCT([mattone_grande,T(3)(z_mattone)(mattone_piccolo),T(3)(z_mattone*2)(mattone_grande),
                    T(3)(z_mattone*3)(mattone_piccolo),T(3)(z_mattone*4)(mattone_grande)])
    mattoni_sx=STRUCT([mattoni,STRUCT(NN(5)([completa,T(3)(z_mattone)])),COMP([T([1,2])([y*0.4,y*0.05]),R([1,2])(PI/2)])(mattoni)])
    mattoni_dx=STRUCT([mattoni,T([1,2])([x*0.1-z_mattone*0.22,y*0.4])(STRUCT(NN(5)([completa,T(3)(z_mattone)]))),COMP([T([1,2])([-y*0.05,y*0.4]),R([1,2])(-PI/2)])(mattoni)])
    
    #assemblo il tutto
    return STRUCT([wall,T(1)(x_wall)(arco),T(1)(x_wall+SIZE(1)(arch)-x*0.1)(wall),
                   T([1,2])([x_wall+SIZE(1)(arch)/2,z*0.5])(scale),muro_sotto2,
                   COMP([T([1,2])([x*24.1,y*0.6]),R([1,2])(PI)])(mattoni_sx),T([1,2])([-x*0,y*0.1])(mattoni_dx)])

#VIEW(parete_nord(1.3,1,1,1))
```

```python
#creo un arco mattonato e due muri a destra e sinistra per creare un muro con un arco mattonato
def bottomArc(d):
    return BEZIER(S1)([[0,0],[0,2*d/3],[d,2*d/3],[d,0]])

def topArc(d):
    return BEZIER(S1)([[0,2*d/3],[d,2*d/3]])

def arc2D(d):
    return BEZIER(S2)([bottomArc(d),topArc(d)])

def arc3D(d):
    def arc3D1(w):
        arco=arc2D(3.2)
        dominio=PROD([INTERVALS(1)(16),INTERVALS(1)(1)])
        ar=MAP(arco)(dominio)
        domin=PROD([ar,QUOTE([1.5])])
        return COMP([T(2)(w),R([2,3])(PI/2)])(domin)
    return arc3D1

def Interarc(d1,d2):
    def Interarc1(w):
        return CUBOID([d1,w,2*d2/3])
    return Interarc1

# lato==1 muro a destra
# lato==0 muro a sinistra
# lunghezza_lato la lunghezza del muro aggiunto
def Xarc(d1,d2):
    def Xarc1(w,lato,lunghezza_lato):
        if lato==1:
            muro=RIGHT([TEXTURE('./image/mattonata.jpg')(Interarc(lunghezza_lato,d2)(w)),
                        TEXTURE('./image/mattonata.jpg')(RIGHT([RIGHT([Interarc(d1,d2)(w),arc3D(d2)(w)]),Interarc(d1,d2)(w)]))])
        else:
            muro=LEFT([TEXTURE('./image/mattonata.jpg')(Interarc(lunghezza_lato,d2)(w)),
                       TEXTURE('./image/mattonata.jpg')(RIGHT([RIGHT([Interarc(d1,d2)(w),arc3D(d2)(w)]),Interarc(d1,d2)(w)]))])
        
        return muro
    return Xarc1

#VIEW(Xarc(0.4,3.2)(1.5,1,0))
```

```python
# crea dei dipinti che stanno su una cupola nel salone del palazzo
def dipinto(x,y,z):
    x=x/1.3
    #definiso punto al centro della cupola e dominio
    curva_up=BEZIER(S1)([[x*2.6,y*2,z*1]]) 
    dominio1=PROD([INTERVALS(1)(32),INTERVALS(1)(1)])
    
    #allungamento dietro
    curva_sx=BEZIER(S1)([[x*0,y*0,z*0],[x*2.6,y*0,z*2],[x*5.2,y*0,z*0]])
    curva_sx_1=BEZIER(S1)([[x*0,-y*1.35,z*0],[x*2.6,-y*1.35,z*2],[x*5.2,-y*1.35,z*0]])
    curva_2D=BEZIER(S2)([curva_sx,curva_sx_1]) 
    curva_2D=MAP(curva_2D)(dominio1)
    curva_2D=TEXTURE('./image/contorno.jpeg')(curva_2D)
    
    #le curve della cupola
    curva_sx2=BEZIER(S1)([[x*0,y*0,z*0],[x*0,y*2,z*2],[x*0,y*4,z*0]])
    curva_2D2=BEZIER(S2)([curva_sx2,curva_up]) 
    curva_2D2=MAP(curva_2D2)(dominio1)
  
    curva_sx3=BEZIER(S1)([[x*0,y*4,z*0],[x*2.6,y*4,z*2],[x*5.2,y*4,z*0]])
    curva_2D3=BEZIER(S2)([curva_sx3,curva_up]) 
    curva_2D3=MAP(curva_2D3)(dominio1)
    
    #allungamneto a sinistra
    curva_sx4=BEZIER(S1)([[x*5.2,y*0,z*0],[x*5.2,y*2,z*2],[x*5.2,y*4,z*0]])
    curva_sx5=BEZIER(S1)([[x*8,y*0,z*0],[x*8,y*2,z*2],[x*8,y*4,z*0]])
    curva_2D5=BEZIER(S2)([curva_sx4,curva_sx5]) 
    curva_2D5=MAP(curva_2D5)(dominio1)
    curva_2D5=TEXTURE('./image/dipinto2.jpeg')(curva_2D5)
    
    #assemblo il tutto
    cupola=TEXTURE('./image/dipinto.jpg')(STRUCT([curva_2D2,curva_2D3,T([1,2])([x*5.2,y*4])(R([1,2])(-PI)(curva_2D2)),
                                                  T([1,2])([x*5.2,y*4])(R([1,2])(PI)(curva_2D3))]))
    return STRUCT([cupola,curva_2D5,curva_2D,T(2)(y*5.35)(curva_2D)])

#VIEW(dipinto(2.6,2.38,1))
```

```python
#colonna crea una colonna
def colonna(x,y,z):
    x=x/1.3
    strato1=COLOR(cMattone)(CUBOID([x,y,z*0.025]))
    strato2=T([1,2])([x*0.1,y*0.1])(COLOR(cMattone)(CUBOID([x*0.8,y*0.8,z*0.015])))
    strato3=T([1,2])([x*0.05,y*0.05])(COLOR(cMattone)(CUBOID([x*0.9,y*0.9,z*0.02])))
    strato4=T([1,2,3])([x*0.1,y*0.1])(COLOR(cMattone)(CUBOID([x*0.8,y*0.8,z*0.04])))
    strato5=T([1,2,3])([x*0.115,y*0.115,z*0.10])(CUBOID([x*0.77,y*0.77,z*0.8]))
    return STRUCT([strato1,T(3)(z*0.025)(strato2),T(3)(z*0.04)(strato3),T(3)(z*0.06)(strato4),strato5,
                  T(3)(z*0.9)(strato4),T(3)(z*0.94)(strato3),T(3)(z*0.96)(strato2),T(3)(z*0.975)(strato1)])
    
#VIEW(colonna(1.3/2,1.0/2,5))
```

```python
#con_tetto==1 con tetto
#con_tetto==0 senza tetto
#crea 3 archi con delle colonne per fare la facciata nord sotto il portico
def muro_archi(x,y,z,con_tetto):
    
    #creo le componenti
    x=x/1.3
    #colonna grande
    colonne_grandi=colonna(x*1.2,y*0.5,z*5)
    x_colonne_grandi=SIZE(1)(colonne_grandi)
    
    #arco
    muro_arco=CUBOID([x*0.4,z*1.5,2*y*3.2/3])
    muro_arco=RIGHT([RIGHT([muro_arco,arc3D(y*3.2)(z*1.5)]),muro_arco])
    z_colonna=SIZE(3)(colonne_grandi)-SIZE(3)(muro_arco)*0.6
    muro_arco=T([1,3])([x*0.7,z_colonna])(S([1,2,3])([x*1.245,y*0.2,z*0.6])(muro_arco))
    x_arco=SIZE(1)(muro_arco)
    
    #colonna piccola
    colonne_piccole=T(1)(x*0.7)(colonna(x*0.64,y*0.4,z_colonna))
    x_colonne_piccole=SIZE(1)(colonne_piccole)
    z_colonne_piccole=SIZE(3)(colonne_piccole)
    y_spostamento=SIZE(2)(muro_arco)
    
    #sporgenza sopra
    sporgenza=T(1)(-x*1.2)(COLOR(cMattone)(CUBOID([x*20,y*0.5,z*0.2])))
    sporgenza=T(3)(SIZE(3)(colonne_grandi))(STRUCT([T(2)(-y*0.2)(sporgenza),T([2,3])([-y*0.125,z*0.2])(sporgenza)]))
    
    #tetto
    if con_tetto==1:
        tetto=T(3)(SIZE(3)(colonne_grandi))(T(1)(-x*1.2)(CUBOID([x*20,y*4.7,z*0.2])))
        sporgenza=STRUCT([sporgenza,T([2,3])([y*0.05,z*0.4])(tetto)])
    
    #assemblo il tutto
    sporgenza_archi=COLOR(cMattone)(T([1,2,3])([x*0.2,y*0.1,z_colonne_piccole-z_colonna*0.1])(CUBOID([x*17,y*0.2,z_colonna*0.1])))
    colonne_grandi=STRUCT([colonne_grandi,T(1)(x_arco+x*0.55)(colonne_grandi),T(1)((x_arco+x*0.57)*2)(colonne_grandi),
                           T(1)((x_arco+x*0.57)*3)(colonne_grandi)])
    colonne_piccole=STRUCT([colonne_piccole,T(1)(x_arco-x*0.5)(colonne_piccole),
                            T(1)(x_arco-x*0.3+x_colonne_grandi)(colonne_piccole),
                            T(1)(x_arco*2+x*0.1)(colonne_piccole),
                            T(1)(x_arco*2+x_colonne_grandi+x*0.28)(colonne_piccole),
                            T(1)(x_arco*3+x*0.7)(colonne_piccole)])
    muro_arco=STRUCT([muro_arco,T(1)(x_arco+x*0.6)(muro_arco),T(1)((x_arco+x*0.6)*2)(muro_arco)])
    
    return STRUCT([colonne_grandi,muro_arco,colonne_piccole,sporgenza,sporgenza_archi])
#VIEW(muro_archi(1.3,1,1,0))
```

```python
#assemblo i 4 muri creati sopra e aggiungo altre componenti elencate sotto
#con_tetto==1 con tetto
#con_tetto==0 senza tetto
#con_secondo_piano==1 con secondo piano
#con_secondo_piano==0 senza secondo piano
#con_tetto==1 con fondamenta
#con_tetto==0 senza fondamenta
def assembla_muri(con_tetto,con_secondo_piano,vedi_fondamenta):
    
    x=1.3
    y=1.0
    z=1.0
    sud=parete_sud(x,y,z)
    muri_ovest=muro_ovest(x,y,z,1,1)
    muri_est=muro_est(x,y,z,con_tetto,con_secondo_piano)
    y_muri=y*0.6
    x_sud=SIZE(1)(sud)
    x_muri=SIZE(1)(muri_ovest)
    muri_ovest=muro_ovest(x,y,z,con_tetto,con_secondo_piano)
    
    #assemblo le pareti
    sud=S(1)(x/1.3*1.003)(COMP([T([1,2])([x_sud-x*0.03,z*0.5]),R([1,2])(PI)])(sud))
    ovest=COMP([T([1,2])([y_muri-0.025*x,-y*0.23+y_muri]),R([1,2])(PI/2)])(muri_ovest)
    est=COMP([T([1,2])([x*23.75,y*21.41]),R([1,2])(-PI/2)])(muri_est)
    nord=T(2)(y*20.78)(parete_nord(x,y,z,con_tetto))
    
    #erba
    pavimento2=T([1,2,3])([-x*0.15+x*0.2,-y*0.3+x*0.2,z*1.8])(CUBOID([x*24.1,x_muri-x*0.67,z*0.2]))
    erba=T([1,2,3])([-x*5,-y*10,z*1.8])(CUBOID([x*35,y*40,z*0.2]))
    erba=T(3)(-z*2)(COLOR(cErba)(DIFFERENCE([erba,pavimento2])))
    
    #pavimento
    pavimento=T([1,2,3])([-x*0.15,-y*0.2,z*1.8])(CUBOID([x*24.4,x*16.7,z*0.2]))
    
    #scale interne
    scale=scala(x*1.2,y*0.3,z*3.7,10,y*0.5*3)(0,0,1)
    scale=T([1,2,3])([x*20-SIZE(1)(scale)-y*0.35,y*12.7,-z*1.7])(scale)
    scale2=T(1)(-x*14.1)(scale)
    pavimento=COLOR(GRAY)(DIFFERENCE([pavimento,T(3)(z*2.8)(scale),T(3)(z*2.8)(scale2)]))
    
    #fondamenta
    x_muro_arco=x*7.818
    fondamenta_arco=Xarc(x*0.4,y*3.2)(z*1.5,0,x_muro_arco)
    z_fondamenta_arco2=SIZE(3)(fondamenta_arco)
    fondamenta_arco=S(3)(z*3.85-z_fondamenta_arco2)(fondamenta_arco)
    x_fondamenta_arco,y_fondamenta_arco,z_fondamenta_arco=SIZE(1)(fondamenta_arco),SIZE(2)(fondamenta_arco),SIZE(3)(fondamenta_arco)
    x_fondamenta_muro=x*7.81
    fondamenta_muro=TEXTURE('./image/mattonata.jpg')(CUBOID([x_fondamenta_muro,y_fondamenta_arco,z_fondamenta_arco]))
    fondamenta_muro=COMP([T([1,2])([x_muro_arco+y_fondamenta_arco,-y*3.5]),R([1,2])(PI/2)])(fondamenta_muro)
    fondamenta_arco_sx=Xarc(x*0.4,y*3.2)(z*1.5,1,x_muro_arco)
    fondamenta_arco_sx=T(1)(x_muro_arco+y_fondamenta_arco)(S(3)(z*3.85-z_fondamenta_arco2)(fondamenta_arco_sx))
    scale_fondamenta=T([1,2])([-x*2.15,-y*3.8])(scala(x,y*0.3,z*1.8,6,y*0.3)(0,0,1))
    fondamenta=STRUCT([fondamenta_arco,T(2)(y*8)(fondamenta_arco),fondamenta_muro,T(2)(x_fondamenta_muro)(fondamenta_muro),
                       fondamenta_arco_sx,T(2)(y*8)(fondamenta_arco_sx),scale_fondamenta,T(1)(x*20.1)(scale_fondamenta)])
    fondamenta=T([1,2,3])([x*3.7,y*4,-z*1.8])(fondamenta)
    
    #affresco
    affresco=dipinto(x*2,y*2.38,z)
    affresco=COMP([T([1,2,3])([x*10.96/1.3,y*16.1,z*9.5]),R([1,2])(-PI/2)])(affresco)
    
    #colonne salone
    colonne=T(3)(z*2)(colonna(x*0.5,y*0.5,z*7.5))
    colonne_angolo_sx=STRUCT([T([1,2])([x*15.69,y*0.2])(colonne),T([1,2])([x*15.6,y*0.1])(colonne)])
    colonne_angolo_dx=STRUCT([T([1,2])([x*8.11,y*0.2])(colonne),T([1,2])([x*8.2,y*0.1])(colonne)])
    colonne=STRUCT([T([1,2])([x*8,y*16])(colonne),T([1,2])([x*15.8,y*16])(colonne),
                    T([1,2])([x*8.1,y*5.4])(colonne),T([1,2])([x*15.7,y*5.5])(colonne),
                    colonne_angolo_dx,T([1,2])([-x*2.3,y*5.3])(colonne_angolo_dx),
                    colonne_angolo_sx,T([1,2])([x*2.3,y*5.4])(colonne_angolo_sx),
                    T([1,2])([x*5.9,y*16])(colonne),T([1,2])([x*17.9,y*16])(colonne),
                    T([1,2])([x*5.8,y*15.9])(colonne),T([1,2])([x*18,y*15.9])(colonne)])
    
    #muro archi
    muro_arco=T([1,2,3])([x*5.3,y*16.4,z*2])(muro_archi(x,y,z,con_tetto))

    #tetto
    if con_tetto==1:
        fondamenta=STRUCT([fondamenta,affresco])
    
    #con fondamenta
    if vedi_fondamenta==1:
        y_scale=SIZE(2)(scale)
        fondamenta=STRUCT([fondamenta,T(3)(-y_scale+y*0.5)(COLOR(GRAY)(pavimento2))])
    
    #assemblo il tutto
    return STRUCT([sud,pavimento,est,erba,ovest,nord,scale,scale2,fondamenta,colonne,muro_arco])

#VIEW della villa completa ci mette qualche secondo a generarsi
#con tetto, con secondo piano e con pavimento fondamenta
VIEW(assembla_muri(1,1,1))

#con tetto, con secondo piano e senza pavimento fondamenta
#VIEW(assembla_muri(1,1,0))

#senza tetto, senza secondo piano e con pavimento fondamenta
#VIEW(assembla_muri(0,0,1))

#senza tetto, con secondo piano e con pavimento fondamenta
#VIEW(assembla_muri(0,1,1))
```

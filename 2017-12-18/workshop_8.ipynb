# Villa Pisani (Bagnolo)

```python
from larlib import *

cLegno = [0.3,0.2,0.1];
cIntonaco = [0.99,0.99,0.99]
cTegola = [0.6,0.05,0]
cMattone = [1,0.95,0.8]
cErba = [0.2,0.4,0.05]
cPietra = [0.7,0.7,0.72]

def hex_material(color, light, trasparence):

    def hex_to_rgb(value):
        value = value.lstrip('#')
        lv = len(value)
        return map(lambda x: x/255., list(int(value[i:i + lv // 3], 16) for i in range(0, lv, lv // 3)))

    params = hex_to_rgb(color) + [.1] + \
             hex_to_rgb(light) + [trasparence] +\
             hex_to_rgb(light) + [.1] +\
             hex_to_rgb("#000000") + [.1] +\
             [100]

    return MATERIAL(params)

window = CUBOID([1,0.1,3])
window = hex_material("#ffffff", "#ffffff", .4)(window)
#VIEW(window)
```

```python
def finestra(x,y,z):
    z_pali=z*0.05
    palo_sopra=COLOR(cLegno)(CUBOID([x,y*0.5,z_pali]))
    
    y=y*0.3
    palo_sotto=COLOR(cLegno)(CUBOID([x,y,z_pali]))
    
    x_palo_orizzontale=x
    z_palo_orizzontale=z*0.05
    palo_orizzontale=COLOR(cLegno)(CUBOID([x_palo_orizzontale,y,z_palo_orizzontale]))
    
    x_palo_verticale=x*0.05
    z_palo_verticale=z*2.95
    palo_verticale=COLOR(cLegno)(CUBOID([x_palo_verticale,y,z_palo_verticale]))
    
    x_vetro_sopra=x*0.45
    z_vetro_sopra=z*1.15
    vetro_sopra=hex_material("#ffffff", "#ffffff", .4)(CUBOID([x_vetro_sopra,y,z_vetro_sopra]))
    
    x_vetro_sotto=x*0.45
    z_vetro_sotto=z*1.7
    vetro_sotto=hex_material("#ffffff", "#ffffff", .4)(CUBOID([x_vetro_sotto,y,z_vetro_sotto]))
    
    x_anta=x*0.05
    ante=COLOR(cLegno)(CUBOID([x_anta,y*2.3,z*3]))
    
    window=STRUCT([ante,T(1)(x_anta)(palo_sotto),
                   T([1,3])([x_anta,z_pali])(vetro_sotto),
                   T([1,3])([x_anta+x_vetro_sotto,z_pali])(palo_verticale),
                   T([1,3])([x_anta+x_vetro_sotto+x_palo_verticale,z_pali])(vetro_sotto),
                   T(1)(x_anta+x_vetro_sotto+x_palo_verticale+x_vetro_sotto)(ante),
                   T([1,3])([x_anta,z_pali+z_vetro_sotto])(palo_orizzontale),
                   T([1,3])([x_anta,z_pali+z_vetro_sotto+z_palo_orizzontale])(vetro_sopra),
                   T([1,3])([x_anta+x_vetro_sopra+x_palo_verticale,z_pali+z_vetro_sotto+z_palo_orizzontale])(vetro_sopra),
                   T([1,3])([x_anta,z_pali+z_vetro_sotto+z_palo_orizzontale+z_vetro_sopra])(palo_sopra)])
    
    z_sporgenza=z*0.05
    sporgenza=COLOR(GRAY)(CUBOID([x*1.2,y*1.6,z_sporgenza]))
    
    return BOTTOM([window,sporgenza])



#VIEW(finestra(1.3,1,1))
```

```python
def finestrella(x,y,z):
    y=y*0.3
    x_contorno,z_contorno=x,z*1.5
    contorno=CUBOID([x_contorno,y*1.95,z_contorno])
    
    x_dif_contorno,z_dif_contorno=x_contorno-0.1,z_contorno-0.1
    dif_contorno=CUBOID([x_dif_contorno,y*1.95,z_dif_contorno])
    
    cornice=DIFFERENCE([contorno,T([1,3])([(x_contorno-x_dif_contorno)/2,(z_contorno-z_dif_contorno)/2])(dif_contorno)])
    cornice=COLOR(cLegno)(cornice)
    
    x_palo=x*0.05
    palo=CUBOID([x_palo,y,z_dif_contorno])
    x_vetro,z_vetro=(x_dif_contorno-x_palo)/2,z_dif_contorno
    vetro=CUBOID([x_vetro,y,z_vetro])
    palo=COLOR(cLegno)(palo)
    vetro=hex_material("#ffffff", "#ffffff", .4)(vetro)
    vetro=STRUCT([vetro,T(1)(x_vetro)(palo),T(1)(x_vetro+x_palo)(vetro)])
    return STRUCT([cornice,T([1,3])([(x_contorno-x_dif_contorno)/2,(z_contorno-z_dif_contorno)/2])(vetro)])

#VIEW(finestrella(1.3,1,1))
```

```python
def finestrella_sotto(x,y,z,tip):
    x_contorno,z_contorno=x,z*1.2
    contorno=CUBOID([x_contorno,y*0.5,z_contorno])
    
    x_dif_contorno,z_dif_contorno=x_contorno-(x*0.1),z_contorno-(z*0.1)
    if tip==0:
        dif_contorno=CUBOID([x_dif_contorno,y*0.5,z_dif_contorno])

        cornice=DIFFERENCE([contorno,T([1,3])([(x_contorno-x_dif_contorno)/2,(z_contorno-z_dif_contorno)/2])(dif_contorno)])
        cornice=COLOR(cLegno)(cornice)

        vetro=hex_material("#ffffff", "#ffffff", .4)(CUBOID([x_dif_contorno,y*0.3,z_dif_contorno]))
        finestra=STRUCT([cornice,T([1,3])([(x_contorno-x_dif_contorno)/2,(z_contorno-z_dif_contorno)/2])(vetro)])
    else:
        x_contorno,z_contorno=x,z*1.5
        contorno=CUBOID([x_contorno,y*0.5,z_contorno])
        dif_contorno=CUBOID([x_dif_contorno,y*0.5,z_contorno-(z*0.05)])
        cornice=DIFFERENCE([contorno,T(1)((x_contorno-x_dif_contorno)/2)(dif_contorno)])
        cornice=COLOR(cLegno)(cornice)
        vetro=hex_material("#ffffff", "#ffffff", .4)(CUBOID([x_dif_contorno,y*0.3,z_contorno-(z*0.05)]))
        finestra=STRUCT([cornice,T(1)((x_contorno-x_dif_contorno)/2)(vetro)])
        
    return finestra

#VIEW(finestrella_sotto(1.3,1,1,1))
```

```python
def porta(x,y,z):
    y=y*0.5
    
    palo_sopra=COLOR(cLegno)(CUBOID([x,y,z*0.05]))
    z_palo_sotto=z*0.15
    palo_sotto=COLOR(cLegno)(CUBOID([x,y,z_palo_sotto]))
      
    x_palo_verticale=x*0.1
    z_palo_verticale=z*2.8
    palo_verticale=COLOR(cLegno)(CUBOID([x_palo_verticale,y,z_palo_verticale]))
    
    x_vetro=x*0.35
    z_vetro=z_palo_verticale
    vetro=hex_material("#ffffff", "#ffffff", .4)(CUBOID([x_vetro,y,z_vetro]))
    
    x_anta=x*0.1
    z_anta=z_vetro
    ante=COLOR(cLegno)(CUBOID([x_anta,y,z_anta]))
    
    door=STRUCT([palo_sotto,T(3)(z_palo_sotto)(ante),
                   T([1,3])([x_anta,z_palo_sotto])(vetro),
                   T([1,3])([x_anta+x_vetro,z_palo_sotto])(palo_verticale),
                   T([1,3])([x_anta+x_vetro+x_palo_verticale,z_palo_sotto])(vetro),
                   T([1,3])([x_anta+x_vetro+x_palo_verticale+x_vetro,z_palo_sotto])(ante),
                   T(3)([z_palo_sotto+z_anta])(palo_sopra)])
    
    return door

#VIEW(porta(1.3,1,1))
```

```python
def scalini(x,y,z,n):
    y_scalino=y*0.5
    z_scalino=z*1.9
    z_abbassa_scalino=z_scalino/n
    
    #n numero scalini
    #i contatore utilizzato anche per abbassare lo scalino
    #y_trasposto di quanto si sposta ogni gradino 
    #scala all'inizio Ã¨ un booleano per creare il primo scalino e poi diventa un oggetto hpc che rappresenta la scala 
    def scalini1(i,y_trasposto,scala):
        if i>=n:
            return scala
        else:
            if scala==1:
                y_scalino_1=y_scalino*5
                scalino=CUBOID([x*7,y_scalino_1,z_scalino])
                lato=CUBOID([x*0.5,y_scalino_1,z_scalino*1.05])
                scalino=RIGHT([lato,RIGHT([scalino,lato])])
                return scalini1(i+1,y_scalino_1,scalino)
            else:
                triangolo=MKPOL([[[0,0],[y_scalino,0],[0,z_abbassa_scalino]],[[1,2,3]],[[]]])
                triangolo=PROD([triangolo,QUOTE([x*0.5])])
                triangolo=COMP([R([1,3])(-PI/2),R([1,2])(PI/2)])(triangolo)
                colonna_lato=TOP([CUBOID([x*0.5,y_scalino,(z_scalino-(z_abbassa_scalino*i))*1.05]),triangolo])
                scalino=CUBOID([x*7,y_scalino,z_scalino-(z_abbassa_scalino*i)])
                scalino=RIGHT([colonna_lato,RIGHT([scalino,colonna_lato])])
                scala=STRUCT([scala,T(2)(y_trasposto)(scalino)])
                return scalini1(i+1,y_trasposto+y_scalino,scala)
    return scalini1

#VIEW(scalini(1.3,1,1,10)(0,0,1))
```

```python
def arco(x,y,z):
    curva_esterna=BEZIER(S1)([[x*0,z*0],[x*0.5,z*2],[x*2.5,z*4],[x*4.5,z*2],[x*5,z*0]])
    curva_interna=BEZIER(S1)([[x*0.2,z*0],[x*0.5,z*1.8],[x*2.5,z*3.6],[x*4.5,z*1.8],[x*4.8,z*0]]) 
    curva_2D=BEZIER(S2)([curva_esterna,curva_interna]) 
    dominio=PROD([INTERVALS(1)(32),INTERVALS(1)(1)])
    curva_2D=MAP(curva_2D)(dominio)
    return PROD([curva_2D,QUOTE([y*0.5])])

def finestra_curva_laterale(x,y,z):
    curva_esterna=BEZIER(S1)([[x*0,z*0],[x*0.42,z*1.24],[x*1.4,z*2.08]])
    curva_interna=BEZIER(S1)([[x*1.4,z*0],[x*1.4,z*2.08]])
    curva_2D=BEZIER(S2)([curva_esterna,curva_interna])
    dominio=PROD([INTERVALS(1)(32),INTERVALS(1)(1)])
    curva_2D=MAP(curva_2D)(dominio)
    return PROD([curva_2D,QUOTE([y*0.5])])

def finestra_centro(x,y,z):
    curva_esterna=BEZIER(S1)([[x*0,z*2.2],[x*0.8,z*2.4],[x*1.6,z*2.2]])
    curva_interna=BEZIER(S1)([[x*0,z*0],[x*1.6,z*0]])
    curva_2D=BEZIER(S2)([curva_esterna,curva_interna])
    dominio=PROD([INTERVALS(1)(32),INTERVALS(1)(1)])
    curva_2D=MAP(curva_2D)(dominio)
    return PROD([curva_2D,QUOTE([y*0.5])])

def finestra_arco(x,y,z):
    finestra_curva_laterale_sx=hex_material("#ffffff", "#ffffff", .4)(JOIN(finestra_curva_laterale(x,y*0.95,z)))
    finestra_curva_laterale_dx=COMP([T([1,3])([x*4.9,y*0.485]),R([1,3])(PI)])(finestra_curva_laterale_sx)
    aste=COLOR(cLegno)(CUBOID([x*0.2,z*2.2,y*0.5]))
    finestra_centrale=finestra_centro(x,y*0.95,z)
    finestra_centrale=hex_material("#ffffff", "#ffffff", .4)(JOIN(finestra_centrale))
    arc=COLOR(cLegno)(arco(x,y,z))
    finestra_senza_asta_sotto=STRUCT([T([1,3])([0.1*x,y*0.013])(finestra_curva_laterale_sx),arc,finestra_curva_laterale_dx,T(1)(x*1.5)(aste),T(1)(x*3.3)(aste),T([1,3])([x*1.7,y*0.013])(finestra_centrale)])
    asta_sotto=COLOR(cLegno)(CUBOID([x*5,z*0.2,y*0.5]))
    return DOWN([finestra_senza_asta_sotto,asta_sotto])

#VIEW(finestra_arco(1,2,1))
```

```python
from larlib import *

def parete_sud(x,y,z):
    #creo muro e faccio buchi
    x_wall=x*24
    wall=CUBOID([x_wall,y*0.5,z*8.5])
    windows=finestra(x,y*(0.5/0.3),z)
    windows=T([1,3])([x*1.5,z*0.7])(windows)
    little_windows=T([1,3])([x*1.5,z*6.5])(finestrella(x,y*(0.5/0.3),z))
    little_windows_dx=STRUCT(NN(2)([little_windows,T(1)(x*4)]))
    little_windows_sx=T(1)(x*16)(little_windows_dx)
    finestra_centrale=finestra_arco(x*(1/1.3),y*2,z)
    door=porta(x,y,z)
    wall=DIFFERENCE([wall,STRUCT(NN(6)([windows,T(1)(x*4)])),little_windows_dx,little_windows_sx,
                     COMP([T([1,2,3])([(x*10.15384),y*0.5*(0.5/0.3),z*5.5]),R([2,3])(PI/2)])(finestra_centrale),
                     T([1,3])([x*11.5,-z*0.2])(door)])
    
    #metto gli oggetti nel muro
    windows=finestra(x,y,z)
    windows=T([1,3])([x*1.5,z*0.7])(windows)
    little_windows=T([1,3])([x*1.5,z*6.5])(finestrella(x,y,z))
    little_windows_dx=STRUCT(NN(2)([little_windows,T(1)(x*4)]))
    little_windows_sx=T(1)(x*16)(little_windows_dx)
    wall=STRUCT([wall,STRUCT(NN(6)([windows,T(1)(x*4)])),little_windows_dx,little_windows_sx])

    finestra_centrale=finestra_arco(x*(1/1.3),y,z)
    
    wall=STRUCT([wall,
                 COMP([T([1,2,3])([(x*10.15384),y*0.5,z*5.5]),R([2,3])(PI/2)])(finestra_centrale),
                 T([1,3])([x*11.5,-z*0.2])(door)])
    x_sporgenza=((x_wall+0.2)/2)-(x*1.4/2)
    sporgenza=COLOR(GRAY)(CUBOID([x_sporgenza,y*0.6,z*0.2]))
    wall=STRUCT([T([1,3])([x*0.1,z*0.2])(wall),sporgenza,T(1)(x_sporgenza+(x*1.4))(sporgenza)])
    
    #faccio muro sotto e buchi
    muro_sotto=CUBOID([x_wall,y*0.5,z*1.8])
    finestra_sotto=finestrella_sotto(x,y*(0.5/0.3),z,0)
    finestra_sotto_1=finestrella_sotto(x,y*(0.5/0.3),z,1)
    muro_sotto=DIFFERENCE([muro_sotto,T([1,3])([x*5.5,z*0.3])(finestra_sotto),T([1,3])([x*17.5,z*0.3])(finestra_sotto),
                           T(1)(x*1.5)(finestra_sotto_1),T(1)(x*21.5)(finestra_sotto_1)])
    
    #metto gli oggetti nel muro sotto
    finestra_sotto=finestrella_sotto(x,y,z,0)
    finestra_sotto_1=finestrella_sotto(x,y,z,1)
    scale=scalini(x,y,z,10)(0,0,1)
    muro_sotto=STRUCT([muro_sotto,T([1,3])([x*5.5,z*0.3])(finestra_sotto),T([1,3])([x*17.5,z*0.3])(finestra_sotto),
                       T(1)(x*1.5)(finestra_sotto_1),T(1)(x*21.5)(finestra_sotto_1),T([1,2])([x*8,y*0.5])(scale)])
    
    
    return STRUCT([T(1)(x*0.1)(muro_sotto),T(3)(z*2)(wall)])

#VIEW(parete_sud(1.3,1,1))
```

```python
def muri_laterali(x,y,z,tip):
    if tip==0:
        x_wall=x*12
        wall=CUBOID([x_wall,y*0.5,z*8.5])
        x_wall2=x*4
        wall2=CUBOID([x_wall2,y*0.5,z*10])
        wall=STRUCT([wall,T(1)(x_wall)(wall2)])
        windows=finestra(x,y*(0.5/0.3),z)
        windows_sotto=T([1,3])([x*1.5,z*0.7])(windows)
        windows_sopra=T([1,3])([x*13.5,z*6.5])(windows)
        little_windows=T([1,3])([x*1.5,z*6.5])(finestrella(x,y*(0.5/0.3),z))
        wall=DIFFERENCE([wall,windows_sopra,STRUCT(NN(4)([windows_sotto,T(1)(x*4)])),STRUCT(NN(3)([little_windows,T(1)(x*4)]))])


        #metto gli oggetti nel muro
        windows=finestra(x,y,z)
        windows_sotto=T([1,3])([x*1.5,z*0.7])(windows)
        windows_sopra=T([1,3])([x*13.5,z*6.5])(windows)
        little_windows=T([1,3])([x*1.5,z*6.5])(finestrella(x,y,z))
        wall=STRUCT([wall,windows_sopra,STRUCT(NN(4)([windows_sotto,T(1)(x*4)])),STRUCT(NN(3)([little_windows,T(1)(x*4)]))])

        x_sporgenza=(x_wall+x_wall2+0.2)
        sporgenza=COLOR(GRAY)(CUBOID([x_sporgenza,y*0.6,z*0.2]))
        wall=STRUCT([T([1,3])([x*0.1,z*0.2])(wall),sporgenza])

        #faccio muro sotto e buchi
        muro_sotto=CUBOID([x_wall+x_wall2,y*0.5,z*1.8])
        finestra_sotto=T([1,3])([x*1.5,z*0.3])(finestrella_sotto(x,y*(0.5/0.3),z,0))
        muro_sotto=DIFFERENCE([muro_sotto,STRUCT(NN(3)([finestra_sotto,T(1)(x*4)]))])

        #metto gli oggetti nel muro sotto
        finestra_sotto=T([1,3])([x*1.5,z*0.3])(finestrella_sotto(x,y,z,0))
        muro_sotto=STRUCT([muro_sotto,STRUCT(NN(4)([finestra_sotto,T(1)(x*4)]))])
        wall=STRUCT([T(1)(x*0.1)(muro_sotto),T(3)(z*2)(wall)])
    #muro piu alto invertito
    else:
        x_wall=x*12
        wall=CUBOID([x_wall,y*0.5,z*8.5])
        x_wall2=x*4
        wall2=CUBOID([x_wall2,y*0.5,z*10])
        wall=STRUCT([wall2,T(1)(x_wall2)(wall)])
        windows=finestra(x,y*(0.5/0.3),z)
        windows_sotto=T([1,3])([x*1.5,z*0.7])(windows)
        windows_sopra=T([1,3])([x*1.5,z*6.5])(windows)
        little_windows=T([1,3])([x*5.5,z*6.5])(finestrella(x,y*(0.5/0.3),z))
        wall=DIFFERENCE([wall,windows_sopra,STRUCT(NN(4)([windows_sotto,T(1)(x*4)])),STRUCT(NN(3)([little_windows,T(1)(x*4)]))])

        #metto gli oggetti nel muro
        windows=finestra(x,y,z)
        windows_sotto=T([1,3])([x*1.5,z*0.7])(windows)
        windows_sopra=T([1,3])([x*1.5,z*6.5])(windows)
        little_windows=T([1,3])([x*5.5,z*6.5])(finestrella(x,y,z))
        wall=STRUCT([wall,windows_sopra,STRUCT(NN(4)([windows_sotto,T(1)(x*4)])),STRUCT(NN(3)([little_windows,T(1)(x*4)]))])

        x_sporgenza=(x_wall+x_wall2+0.2)
        sporgenza=COLOR(GRAY)(CUBOID([x_sporgenza,y*0.6,z*0.2]))
        wall=STRUCT([T([1,3])([x*0.1,z*0.2])(wall),sporgenza])

        #faccio muro sotto e buchi
        muro_sotto=CUBOID([x_wall+x_wall2,y*0.5,z*1.8])
        finestra_sotto=T([1,3])([x*5.5,z*0.3])(finestrella_sotto(x,y*(0.5/0.3),z,0))
        muro_sotto=DIFFERENCE([muro_sotto,STRUCT(NN(3)([finestra_sotto,T(1)(x*4)]))])

        #metto gli oggetti nel muro sotto
        finestra_sotto=T([1,3])([x*1.5,z*0.3])(finestrella_sotto(x,y,z,0))
        muro_sotto=STRUCT([muro_sotto,STRUCT(NN(4)([finestra_sotto,T(1)(x*4)]))])
        wall=STRUCT([T(1)(x*0.1)(muro_sotto),T(3)(z*2)(wall)])
    
    return wall

#VIEW(muri_laterali(1.3,1,1,1))
```

```python
#tip==1 muro piu alto
#tip==0 muro piu basso

def muro_con_porta(x,y,z,tip):
    if tip==1:
        x_muro,z_muro=x*1.5,z*10
    else:
        x_muro,z_muro=x*1.5,z*8.5
    muro=CUBOID([x_muro,y*0.5,z_muro])
    door=porta(x,y,z)
    z_door=SIZE(3)(door)
    x_muro_porta,z_muro_porta=x*1,z_muro-z_door
    muro_porta=CUBOID([x_muro_porta,y*0.5,z_muro_porta])
    
    
    wall=STRUCT([muro,T([1,3])([x_muro,z_door])(muro_porta),T(1)(x_muro+x_muro_porta)(muro),T(1)(x_muro)(door)])
      
    return wall

def vertici(x,y,z):
    x_pilastro=x
    y_pilastro=y
    z_pilastro=z*0.4
    x_pilastro2=x_pilastro/2
    z_pilastro2=(x_pilastro2/4.1)/COS(PI/6)
    V=[[0,0,0],#1 quadrato dx
    [0,y_pilastro,0],#2 quadrato dx
    [0,y_pilastro,z_pilastro],#3 quadrato dx
    [0,0,z_pilastro],#4 quadrato dx
    [-y_pilastro,y_pilastro*2,z_pilastro],#5 triangolo dx
    [x_pilastro,0,0],#6 quadrato sx
    [x_pilastro,y_pilastro,0],#7 quadrato sx
    [x_pilastro,y_pilastro,z_pilastro],#8 quadrato sx
    [x_pilastro,0,z_pilastro],#9 quadrato sx
    [x_pilastro+y_pilastro,y_pilastro*2,z_pilastro],#10 triangolo sx
    [0,0,-z_pilastro],#11 sotto dx
    [0,y_pilastro,-z_pilastro],#12 sotto dx
    [x_pilastro,0,-z_pilastro],#13 sotto sx
    [x_pilastro,y_pilastro,-z_pilastro],#14 sotto sx
    [x_pilastro2,-(x_pilastro2-(y_pilastro)),z_pilastro2+z_pilastro],# 15 tetto
    ]
    return V

#tip==1 muro ad est
#tip==0 muro ad ovest
#con_tetto==1 con tetto
#con_tetto==0 senza tetto
def interni(x,y,z,tip,con_tetto):
    wall=T(3)(z*2.2)(muro_con_porta(x,y,z,1))
    x_wall=SIZE(1)(wall)
    y_wall=SIZE(2)(wall)
    V=vertici(x_wall*1.02,y_wall*1.1,z*0.5)
    pilastro_rettangolo=MKPOL([V,[[1,4,3,5,2],[6,9,8,10,7]],[]])
    pilastro_rettangolo_cubo=MKPOL([V,[[1,2,11,12],[6,7,13,14]],[]])
    su=MKPOL([V,[[15,10,5]],[]])
    pilastro_rettangolo=COLOR(cTegola)(JOIN(pilastro_rettangolo))
    su=COLOR(cTegola)(su)
    pilastro_rettangolo_cubo=JOIN(pilastro_rettangolo_cubo)
    tetto=STRUCT([pilastro_rettangolo_cubo,pilastro_rettangolo,su])
    tetto_dx=COMP([T([1,2])([y_wall*1.1,-x_wall*1.02+y_wall*1.1]),R([1,2])(PI/2)])(tetto)
    tetto_su=COMP([T([1,2])([x_wall*1.02,-x_wall*1.02+y_wall*1.1*2]),R([1,2])(PI)])(tetto)
    tetto_sx=COMP([T([1,2])([x_wall*1.02-y_wall*1.1,y_wall*1.1]),R([1,2])(-PI/2)])(tetto)
    tetto=STRUCT([tetto,tetto_dx,tetto_su,tetto_sx])
    
    if tip==1:
        muro_laterale=muri_laterali(x,y,z,1)
        
        wall2=COMP([T([1,2])([x_wall+x*0.1,-x_wall]),R([1,2])(PI/2)])(wall)
        
        wall3=T(3)(z*2.2)(muro_con_porta(x*2,y,z,0))
        x_wall3=SIZE(1)(wall3)
        
        wall4=T(3)(z*2.2)(muro_con_porta(x,y,z,0))
        wall4=COMP([T([1,2])([x_wall+x*0.1+x_wall3,-x_wall]),R([1,2])(PI/2)])(wall4)
        
        wall5=T(3)(z*2.2)(muro_con_porta(x,y,z,0))
        
        x_wall6=x*4
        wall6=T(3)(z*2.2)(CUBOID([x_wall6,y*0.5,z*8.5]))
        wall6=COMP([T([1,2])([x_wall+x*0.1+x_wall3,-x_wall*2]),R([1,2])(PI/2)])(wall6)
        
        wall7=T(3)(z*2.2)(muro_con_porta(x*4,y,z,0))
        wall7=COMP([T([1,2])([x_wall+x*0.1,-x_wall-SIZE(1)(wall7)]),R([1,2])(PI/2)])(wall7)
        
        pareti=STRUCT([muro_laterale,T([1,2])([x*0.1,-y*x_wall])(wall),
                       wall2,T([1,2])([x_wall+x*0.1,-y*x_wall])(wall3),wall4,
                       T([1,2])([x_wall+x*0.1+x_wall3,-y*x_wall-x_wall6])(wall5),
                       wall6,wall7])
        
        if con_tetto==1:
            V=vertici((x_wall*3*1.02)+SIZE(2)(wall),y_wall*1.1,z*0.5)
            pilastro_rettangolo=MKPOL([V,[[1,4,3,5,2],[6,9,8,10,7]],[]])
            pilastro_rettangolo_cubo=MKPOL([V,[[1,2,11,12],[6,7,13,14]],[]])
            su=MKPOL([V,[[15,10,5]],[]])
            
            pilastro_rettangolo=COLOR(cTegola)(JOIN(pilastro_rettangolo))
            su=COLOR(cTegola)(su)
            pilastro_rettangolo_cubo=JOIN(pilastro_rettangolo_cubo)
            
            x_sposta_tetto,z_sposta_tetto=x_wall3/2,SIZE(3)(wall3)+z*2.2+SIZE(3)(pilastro_rettangolo_cubo)
            pilastro_rettangolo_cubo=T([1,3])([x_sposta_tetto,z_sposta_tetto])(pilastro_rettangolo_cubo)
            pilastro_rettangolo=T([1,3])([x_sposta_tetto,z_sposta_tetto])(pilastro_rettangolo)
            su=T([1,3])([x_sposta_tetto,z_sposta_tetto])(su)
            #---------------------------------------------------------------------------------------
            x_sposta,y_sposta=SIZE(1)(pilastro_rettangolo_cubo)+SIZE(1)(wall)*2,-SIZE(1)(wall)*6+SIZE(2)(pilastro_rettangolo_cubo)+SIZE(2)(pilastro_rettangolo_cubo)/2
            pilastro_rettangolo_cubo_opposto=COMP([T([1,2])([x_sposta,y_sposta]),R([1,2])(PI)])(pilastro_rettangolo_cubo)
            pilastro_rettangolo_opposto=COMP([T([1,2])([x_sposta,y_sposta]),R([1,2])(PI)])(pilastro_rettangolo)
            su_opposto=COMP([T([1,2])([x_sposta,y_sposta]),R([1,2])(PI)])(su)
            
            pilastro_rettangolo_cubo=JOIN([pilastro_rettangolo_cubo,pilastro_rettangolo_cubo_opposto])
            pilastro_rettangolo=COLOR(cTegola)(JOIN([pilastro_rettangolo_opposto,pilastro_rettangolo]))
            su=COLOR(cTegola)(JOIN([su,su_opposto]))
            
            pareti=STRUCT([pareti,pilastro_rettangolo_cubo,pilastro_rettangolo,su])
            #--------------------------------------------------------------------------------------------
            x_spostamento=(x*0.08+x_wall+x_wall3)-(((x_wall*1.02)-x_wall)/2)
            y_spostamento=y_wall*1.1
            z_spostamento=SIZE(3)(muro_laterale)+z*0.5*0.4            
            pareti=STRUCT([pareti,T(3)(z_spostamento)(S(2)((x_wall*1.02+y_wall*1.1)/x_wall*1.02)(tetto))])

        return pareti
    
    else:
        muro_laterale=muri_laterali(x,y,z,0)
        
        wall3=T(3)(z*2.2)(muro_con_porta(x*2,y,z,0))
        x_wall3=SIZE(1)(wall3)
        
        wall2=COMP([T([1,2])([x_wall+x*0.1+y*0.5+x_wall3,-x_wall]),R([1,2])(PI/2)])(wall)
        
        wall4=T(3)(z*2.2)(muro_con_porta(x,y,z,0))
        wall4=COMP([T([1,2])([x_wall+x*0.1,-x_wall]),R([1,2])(PI/2)])(wall4)
        
        wall5=T(3)(z*2.2)(muro_con_porta(x,y,z,0))
        
        x_wall6=x*4
        wall6=T(3)(z*2.2)(CUBOID([x_wall6,y*0.5,z*8.5]))
        wall6=COMP([T([1,2])([x_wall+x*0.1,-x_wall*2]),R([1,2])(PI/2)])(wall6)
        
        pareti=STRUCT([muro_laterale,T([1,2])([x*0.08+x_wall+x_wall3,-y*x_wall])(wall),
                       wall2,T([1,2])([x_wall+x*0.1,-y*x_wall])(wall3),wall4,
                       T([1,2])([x*0.1,-y*x_wall-x_wall6])(wall5),
                       wall6])
        if con_tetto==1:
            x_spostamento=(x*0.08+x_wall+x_wall3)-(((x_wall*1.02)-x_wall)/2)
            y_spostamento=y_wall*1.1
            z_spostamento=SIZE(3)(pareti)+z*0.5*0.4            
            pareti=STRUCT([pareti,T([1,2,3])([x_spostamento,y_spostamento/8,z_spostamento])(S(2)((x_wall*1.02+y_wall*1.1)/x_wall*1.02)(tetto))])
            
        return pareti

#VIEW(interni(1.3,1,1,1,1))
```

```python
from larlib import *
def ArchSurface(rr,w,h):
    x1 = lambda k: rr*cos(k[0])
    y1 = lambda k: rr*sin(k[0])
    x2 = lambda k: (rr-w)*cos(k[0])
    y2 = lambda k: (rr-w)*sin(k[0])
    z = lambda k: h
    
    Circle0 = BEZIER(S1)([CONS([x1,y1,z])])
    Circle1 = BEZIER(S1)([CONS([x2,y2,z])])
    return BEZIER(S2)([Circle0, Circle1])

# w larghezza dei cubi
# length lungezza del'arco
# depth profonditÃ  dei cubi
def Arch(length,w,depth):
    def Arch1(angle):
        
        radius=(length/2)/cos(angle/2)
        domain2D=PROD([T(1)(angle/2)(INTERVALS(PI-angle)(13)),QUOTE([1])])
        domain3D=PROD([domain2D, QUOTE([1])])
        ArchSurf2D_0=ArchSurface(radius,w,0)
        ArchSurf2D_1=ArchSurface(radius,w,depth)
        SolidMap=BEZIER(S3)([ArchSurf2D_0, ArchSurf2D_1])
        return MAP(SolidMap)(domain3D)
    return Arch1

myArch = Arch(10, 2, 1)(0.001*PI)
#VIEW(myArch)
```

```python
def colonnina(x,y,z):
    x=x/1.3
    curva_sx=BEZIER(S1)([[x*1,y*0,z*0],[-x*2,y*0,z*1],[x*5,y*0,z*2],[-x*2,y*0,z*3],[x*1,y*0,z*4]])
    curva_up=BEZIER(S1)([[x*0,-y*1,z*0],[x*0,y*2,z*1],[x*0,-y*5,z*2],[x*0,y*2,z*3],[x*0,-y*1,z*4]]) 
    curva_2D=BEZIER(S2)([curva_sx,curva_up]) 
  
    curva_dx=BEZIER(S1)([[-x*1,y*0,z*0],[x*2,y*0,z*1],[-x*5,y*0,z*2],[x*2,y*0,z*3],[-x*1,y*0,z*4]]) 
    curva_down=BEZIER(S1)([[x*0,y*1,z*0],[x*0,-y*2,z*1],[x*0,y*5,z*2],[x*0,-y*2,z*3],[x*0,y*1,z*4]])
    curva_2D1=BEZIER(S2)([curva_dx,curva_down]) 
    
    dominio1=PROD([INTERVALS(1)(32),INTERVALS(1)(1)])

    curva_3D=BEZIER(S3)([curva_2D,curva_2D1]) 
    dominio3d=PROD([dominio1,QUOTE([1])])
    curva_3D=MAP(curva_3D)(dominio3d)
    return COLOR(cMattone)(curva_3D)

#VIEW(colonnina(1.3,1,1))
```

```python
def muretto(x,y,z):
    
    col=R([1,2])(PI/4)(colonnina(x,y,z))
    x_dim_col,y_dim_col=SIZE(1)(col),SIZE(2)(col)
    muretto=STRUCT(NN(5)([col,T(1)((x_dim_col)+(x_dim_col)/10)]))
    mur_centro=T([1,2])([-x_dim_col/2,-y_dim_col/2])(CUBOID([x_dim_col,y_dim_col,SIZE(3)(col)]))
    muretto=STRUCT([muretto,T(1)(SIZE(1)(muretto)+(x_dim_col)/10)(mur_centro),
                   T(1)(SIZE(1)(mur_centro)+SIZE(1)(muretto)+(x_dim_col)*2/10)(muretto)])
    x_asta=SIZE(1)(muretto)*1.1
    z_asta=SIZE(3)(muretto)/10
    asta=CUBOID([x_asta,SIZE(2)(muretto),z_asta])
    muretto=T([1,2,3])([(x_dim_col/2)+(x_asta-(x_asta/1.1))/2,y_dim_col/2,z_asta])(TOP([BOTTOM([muretto,asta]),asta]))
    
    return COLOR(cMattone)(muretto)

#VIEW(muretto(1.3,1,1))
```

```python
def archi(x,y,z):
    depth=y*0.5
    x_mattone=x*0.8
    z_mattone=z*0.3625
    n=11
    
    #creo arco dx
    mattone=CUBOID([x_mattone,depth,z_mattone])
    colonna=STRUCT(NN(n)([mattone,T(3)(z*0.3625)]))
    lunghezza_arco=1.3*0.8*3.75
    w=1.3*0.8
    arch=Arch(lunghezza_arco, w , 0.5)(0)
    arco=S([1,2,3])([x,z,y])(arch)
    arco=STRUCT([T(1)(-(1.3/(lunghezza_arco+w)))(colonna),T(1)(lunghezza_arco+(1.3/(lunghezza_arco+w))-x_mattone)(colonna),
                 COMP([T([1,2,3])([x*1.5,depth,z_mattone*n]),R([2,3])(PI/2)])(arco)])
    #creo colonna
    x_mattone2=x*0.8
    y_mattone2=y*0.52
    mattone2=CUBOID([x_mattone2,y_mattone2,z_mattone])
    x_mattone_sopra=x_mattone2/2
    mattone_sopra=CUBOID([x_mattone_sopra,y_mattone2,z_mattone])
    colonna=STRUCT(NN(16)([mattone2,T(3)(z_mattone)]))
    colonna_sx=STRUCT([T([1,3])([x_mattone2,z_mattone*15])(mattone_sopra),colonna])
    colonna_dx=RIGHT([colonna,colonna_sx])
    arco_dx=STRUCT([T(1)(x_mattone2+x_mattone2*0.5)(arco),T(2)(z*-0.01)(colonna_dx)])
    
    #creo arco centro
    colonna_centro=STRUCT([T([1,3])([x_mattone2,z_mattone*15])(mattone_sopra),colonna,
                           T([1,3])([-x_mattone_sopra,z_mattone*15])(mattone_sopra)])
    arco_centro=STRUCT([T(1)(x_mattone2*0.5)(arco),T(2)(z*-0.01)(colonna_centro),
                        T([1,2])([(x/1.3)*lunghezza_arco,z*-0.01])(colonna_centro)])
    
    #creo arco sinistra
    colonna_sxx=STRUCT([T([1,3])([-x_mattone_sopra,z_mattone*15])(mattone_sopra),colonna])
    colonna_dxx=LEFT([colonna,colonna_sxx])
    arco_sx=STRUCT([arco,T([1,2])([(x/1.3)*lunghezza_arco+x_mattone2*0.5,z*-0.01])(colonna_dxx)])
    archi=STRUCT([arco_dx,T(1)((x/1.3)*lunghezza_arco+x_mattone2)(arco_centro),
                                   T(1)((x/1.3)*lunghezza_arco*2+x_mattone2+x_mattone2*0.5)(arco_sx)])
    asta_sopra=CUBOID([SIZE(1)(archi),y_mattone2,z*0.5])
    archi=STRUCT([archi,T([2,3])([-0.01,z_mattone*16])(asta_sopra)])
    x1=x/1.3
    muro_colonnine=S([1,2,3])([x1*0.128,y*0.36,z*0.2])(muretto(x,y,z))
    archi=STRUCT([archi,T(1)(x_mattone2*2+x_mattone2/4)(muro_colonnine),
                  T(1)(SIZE(1)(archi)-(x_mattone2*2+x_mattone2/4)-SIZE(1)(muro_colonnine))(muro_colonnine)])
    return S([1,3])([(x/1.2975)*1.41,z*0.9])(COLOR(cMattone)(archi))

#VIEW(archi(1.3,1,1))
```

```python
#con_tetto se Ã¨ 1 c'Ã¨ anche il tetto se no non ci sta
def pilastro_tetto(x,y,z,con_tetto):
    
    x_pilastro=x*1.01
    y_pilastro=y*1.05
    z_pilastro=z*0.4
    x_pilastro2=x_pilastro/2
    z_pilastro2=(x_pilastro2/4.1)/COS(PI/6)
    
    V=[[0,0,0],#1 quadrato dx
    [0,y_pilastro,0],#2 quadrato dx
    [0,y_pilastro,z_pilastro],#3 quadrato dx
    [0,0,z_pilastro],#4 quadrato dx
    [0,y_pilastro*2,z_pilastro],#5 triangolo dx
    [x_pilastro,0,0],#6 quadrato sx
    [x_pilastro,y_pilastro,0],#7 quadrato sx
    [x_pilastro,y_pilastro,z_pilastro],#8 quadrato sx
    [x_pilastro,0,z_pilastro],#9 quadrato sx
    [x_pilastro,y_pilastro*2,z_pilastro],#10 triangolo sx
    [x_pilastro2,0,z_pilastro2],#11 quadrato su
    [x_pilastro2,y_pilastro,z_pilastro2],#12 quadrato su
    [x_pilastro2,y_pilastro,z_pilastro2+z_pilastro],#13 quadrato su
    [x_pilastro2,0,z_pilastro2+z_pilastro],#14 quadrato su
    [x_pilastro2,y_pilastro*2,z_pilastro2+z_pilastro],#15 triangolo su
    [0,0,-z_pilastro],#16 sotto dx
    [0,y_pilastro,-z_pilastro],#17 sotto dx
    [x_pilastro,0,-z_pilastro],#18 sotto sx
    [x_pilastro,y_pilastro,-z_pilastro],#19 sotto sx
    [x_pilastro2,0,z_pilastro2-z_pilastro], #20 sotto su
    [x_pilastro2,y_pilastro,z_pilastro2-z_pilastro], #21 sotto su
    [x_pilastro2,-(x/(1.3*3.9)),z_pilastro2-(z_pilastro*1.8)],# 22 tetto
    [y_pilastro,-(x/(1.3*3.9)),z_pilastro2-(z_pilastro*1.8)],# 23 tetto
    [x_pilastro-y_pilastro,-(x/(1.3*3.9)),z_pilastro2-(z_pilastro*1.8)],# 24 tetto
    [y_pilastro,0,z_pilastro],# 25 tetto
    [x_pilastro-y_pilastro,0,z_pilastro]# 26 tetto
    ]

    pilastro_angolo_sx=MKPOL([V,[ [6,9,8,10,7],[11,14,13,15,12]],[]])    
    pilastro_angolo_sx_cubo=MKPOL([V,[[6,7,18,19],[11,12,20,21]],[]])
    pilastro_angolo_dx=MKPOL([V,[[1,4,3,5,2],[11,14,13,15,12]],[]])
    pilastro_angolo_dx_cubo=MKPOL([V,[[1,2,16,17],[11,12,20,21]],[]])
    pilastro_rettangolo=MKPOL([V,[[1,4,3,5,2],[6,9,8,10,7]],[]])
    pilastro_rettangolo_cubo=MKPOL([V,[[1,2,16,17],[6,7,18,19]],[]])
    sfondo=MKPOL([V,[[4,9,11]],[]])
    
    pilastro_angolo_sx=JOIN(pilastro_angolo_sx)
    pilastro_angolo_sx_cubo=JOIN(pilastro_angolo_sx_cubo)
    pilastro_angolo_dx=JOIN(pilastro_angolo_dx)
    pilastro_angolo_dx_cubo=JOIN(pilastro_angolo_dx_cubo)
    pilastro_rettangolo=JOIN(pilastro_rettangolo)
    pilastro_rettangolo_cubo=JOIN(pilastro_rettangolo_cubo)
    sfondo=COMP([T(1)(x_pilastro),R([1,2])(PI)])(JOIN(sfondo))
    pilastri=STRUCT([sfondo,pilastro_angolo_sx,pilastro_angolo_sx_cubo,pilastro_angolo_dx,
            pilastro_angolo_dx_cubo,pilastro_rettangolo,pilastro_rettangolo_cubo])
    
    if con_tetto==1:
        pilastro_angolo_sx=COLOR(cTegola)(pilastro_angolo_sx)
        pilastro_angolo_dx=COLOR(cTegola)(pilastro_angolo_dx)
        tetto=COLOR(cTegola)(STRUCT([JOIN(MKPOL([V,[[14,22,26]],[]])),MKPOL([V,[[14,25,22]],[]]),
                                     JOIN(MKPOL([V,[[22,23,25]],[]])),JOIN(MKPOL([V,[[26,24,22]],[]]))]))
        tetto=STRUCT([tetto,pilastro_angolo_sx,pilastro_angolo_sx_cubo,pilastro_angolo_dx,pilastro_angolo_dx_cubo])
        tetto=T(3)(SIZE(3)(pilastro_rettangolo)+SIZE(3)(pilastro_rettangolo_cubo))(tetto)
        pilastri=STRUCT([pilastri,tetto])
    
    return pilastri

#VIEW(pilastro_tetto(20.9364624023,0.519999980927,1,1))
```

```python
def scala_cilidriche(x,y,z):

    x1= lambda (u,v) : v*cos(u) 
    y1= lambda (u,v) : v*sin(u)

    dom= PROD([INTERVALS(PI)(32),INTERVALS(1)(32)])
    mapp= MAP([x1,y1])(dom)
    
    return JOIN(S([1,2])([(x/1.3),y])(PROD([mapp,QUOTE([z*2.0/6.0])])))

def scale_cilindriche(x,y,z):
    # n numero scalini
    # scala prima Ã¨ 1 per il primo scalino poi diventa un oggetto hpc scala
    # i serve per mettere gli scalini uno sopra l'altro
    def scalini(n,scala,i):
        if n<1 :
            return scala
        else:
            scalino=scala_cilidriche(x*(n+2),y*(n+2),z)
            if scala==1:   
                return scalini(n-1,scalino,i+2.0/6.0)
            else:
                scala=STRUCT([scala,T(3)(i)(scalino)])
                return scalini(n-1,scala,i+2.0/6.0)
    return scalini
                

#VIEW(scale_cilindriche(1.3,1,1)(6,1,0))
```

```python

def parete_nord(x,y,z):  
    #faccio muro
    x_wall=x*4
    wall=CUBOID([x_wall,y*0.5,z*10])
    windows=finestra(x,y*(0.5/0.3),z)
    windows=T([1,3])([x*1.5,z*0.7])(windows)
    wall=DIFFERENCE([wall,windows,T(3)(z*5.8)(windows)])

    #metto gli oggetti nel muro
    windows=finestra(x,y,z)
    windows=T([1,3])([x*1.5,z*0.7])(windows)
    wall=STRUCT([wall,windows,T(3)(z*5.8)(windows)])
    x_sporgenza=(x_wall+0.2)
    sporgenza=COLOR(GRAY)(CUBOID([x_sporgenza,y*0.6,z*0.2]))
    wall=STRUCT([T([1,3])([x*0.1,z*0.2])(wall),sporgenza])
    
    #faccio muro sotto e buchi
    z_muro_sotto=z*1.8
    muro_sotto=CUBOID([x_wall,y*0.5,z_muro_sotto])
    finestra_sotto=T([1,3])([x*1.5,z*0.3])(finestrella_sotto(x,y*(0.5/0.3),z,0))
    muro_sotto=DIFFERENCE([muro_sotto,finestra_sotto])
    
    #metto gli oggetti nel muro sotto
    finestra_sotto=T([1,3])([x*1.5,z*0.3])(finestrella_sotto(x,y,z,0))
    muro_sotto=STRUCT([muro_sotto,finestra_sotto])
    
    wall=STRUCT([T(1)(x*0.1)(muro_sotto),T(3)(z_muro_sotto+(z*0.2))(wall)])
    
    arch=archi(x,y,z)   
    pilastro=pilastro_tetto(SIZE(1)(arch),SIZE(2)(arch),z,1)
    arco=STRUCT([arch,T([1,3])([(-SIZE(1)(arch)*(1.005))+SIZE(1)(arch),SIZE(3)(arch)+z*0.4])(pilastro)])
    
    sotto_archi=COLOR(cMattone)(CUBOID([SIZE(1)(arco),y*0.5,z_muro_sotto]))
    arco=STRUCT([sotto_archi,T(3)(z_muro_sotto+(z*0.2))(arco)])
    
    scale=scale_cilindriche(x/1.87,y/1.87,z)(6,1,0)

    return STRUCT([wall,T(1)(x_wall)(arco),T(1)(x_wall+SIZE(1)(arch)-x*0.1)(wall),T([1,2])([x_wall+SIZE(1)(arch)/2,z*0.5])(scale)])

#VIEW(parete_nord(1.3,1,1))
```
